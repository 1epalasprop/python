# Εκφώνηση

Μια ομάδα μπάσκετ έδωσε στην Euroleage 12 αγώνες.

Να αναπτύξετε πρόγραμμα στη γλώσσα Python το οποίο:

1. Να διαβάζει τους πόντους που πέτυχε η ομάδα σε κάθε αγώνα. Οι πόντοι αυτοί
   πρέπει να είναι θετικός αριθμός.
1. Για κάθε αγώνα, να εμφανίζει το μήνυμα «Υπέροχα» αν η επίδοση της ομάδας είναι
   άνω των 85 πόντων.
1. Να εμφανίζει το μέσο όρο των πόντων που πέτυχε η ομάδα στους 12 αυτούς αγώνες.

## 1ο Ζητούμενο

Το 1ο ζητούμενο έχει δύο (2) μέρη:
1. Να διαβάζει τους πόντους **σε κάθε αγώνα**
2. Να ελέγχει ότι οι πόντοι είναι **θετικός αριθμός**

Για να διαβάσουμε τους πόντους σε κάθε αγώνα, **αφού γνωρίζουμε πόσους αγώνες**
έχει, χρησιμοποιούμε τη `for`.

```python
for match in range(12):
    points = int(input("Δώσε τους πόντους που αγώνα: "))
```

Για να ελέγξουμε ότι η μεταβλητή `points` θα έχει θετικό αριθμό, χρησιμοποιούμε
*έλεγχο εγκυρότητας* με τη `while`.

```python
for ...:
    points = ...
    while points < 0:  # Όσο οι πόντοι είναι αρνητικοί
        print("Λάθος τιμή")  # Προαιρετικό αλλά χρήσιμο
        points = int(input("Δώσε τους πόντους του αγώνα: "))
```

## 2ο Ζητούμενο

Στο ζητούμενο λέει "**σε κάθε αγώνα** να εμφανίζει το μήνυμα". Αυτό, σημαίνει
ότι ο έλεγχος και το μήνυμα θα πρέπει να είναι **μέσα** στη `for`.

```python
for ...:
    points = ...
    while ...:
        print...
        points = ...
    # Εμφάνιση μηνύματος μόνο αν οι πόντοι είναι πάνω από 85.
    if points > 85:
        print("Υπέροχα")
```

## 3ο Ζητούμενο

Για να υπολογίσουμε τον μέσο όρο, χρειάζεται να υπολογίσουμε το **άθροισμα όλων των πόντων**.
Χρειαζόμαστε μία μεταβλητή γι' αυτό το σκοπό και θα πρέπει να πάρει αρχική τιμή **πρίν**
από τη `for` γιατί **αφορά όλους τους αγώνες**.

Τον μέσο όρο μπορούμε να τον υπολογίζουμε **μόνο όταν έχουμε ολοκληρώσει τη `for`**.
Αυτό σημαίνει ότι ο υπολογισμός και η εμφάνιση θα γίνει **μετά** τη `for`.

Για τον υπολογισμό του μέσου όρου χρειαζόμαστε δύο όρους: *α)* τον αριθμητή,
που είναι το **άθροισμα**, και *β)* τον παρανομαστή που είναι το **πλήθος**
των επαναλήψεων που έχουμε κάνει. Το άθροισμα το υπολογίζουμε μέσα στη `for`.
Το πλήθος το ξέρουμε, αφού έδωσε 12 αγώνες.

```python
athroisma = 0  # Συνολικό άθροισμα πόντων
for ...:
    points = ...
    while ...:
        print...
        points = ...
    if ...:
        print...
    # Πρόσθεσε τους πόντους του αγώνα (είμαστε μέσα στη for) στο άθροισμα
    athroisma = athroisma + points
# Έξω από τη for
mesos_oros = float(athroisma) / 12  # 12 είναι οι επαναλήψεις του for
print("Ο μέσος όρος των πόντων της ομάδας είναι: ", mesos_oros)
```

# Ολοκληρωμένο Πρόγραμμα

````python
athroisma = 0  # Συνολικό άθροισμα πόντων
for match in range(1, 13):
    print("Αγώνας", match)
    points = int(input("Δώσε τους πόντους του αγώνα: "))
    while points < 0:
        print("Λάθος τιμή")
        points = int(input("Δώσε τους πόντους του αγώνα: "))
    if points > 85:
        print("Υπέροχα")
    # Πρόσθεσε τους πόντους του αγώνα (είμαστε μέσα στη for) στο άθροισμα
    athroisma = athroisma + points
# Έξω από τη for
mesos_oros = float(athroisma) / 12  # 12 είναι οι επαναλήψεις του for
print("Ο μέσος όρος των πόντων της ομάδας είναι: ", mesos_oros)
````
