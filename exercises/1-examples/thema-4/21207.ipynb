# Εκφώνηση

Στον διαγωνισμό Πληροφορικής Bebras συμμετέχουν δέκα (10) μαθητές. Κάθε μαθητής
που συμμετέχει λαμβάνει μια τελική βαθμολογία από 1 έως και 100 ακέραιες μονάδες.

Να γράψετε πρόγραμμα σε γλώσσα προγραμματισμού Python, το οποίο:

1. Να διαβάζει το ονοματεπώνυμο κάθε μαθητή και τη βαθμολογία που έλαβε.
   Να γίνει έλεγχος ορθότητας ως προς τη βαθμολογία. Τα στοιχεία αυτά
   καταχωρίζονται σε δύο λίστες.
3. Να υπολογίζει και να εμφανίζει τον μέσο όρο της βαθμολογίας όλων των μαθητών. 
4. Να εντοπίζει και να εμφανίζει στην οθόνη τα ονοματεπώνυμα και τη βαθμολογία
   όλων των μαθητών των οποίων η τελική βαθμολογία είναι μεγαλύτερη ή ίση από τον μέσο όρο. 


# 1ο Ζητούμενο

Το 1ο ζητούμενο έχει τρία μέρη:

1. Να διαβάζει το ονοματεπώνυμο και τη βαθμολογία του μαθητή.
2. Να ελέγχει την βαθμολογία.
3. Να εισάγει τα δεδομένα σε δύο πίνακες.

Γνωρίζουμε ότι έχουμε 10 μαθητές, άρα μπορούμε να χρησιμοποιήσουμε τη δομή 
επανάληψης `for`.

```python
for i in range(10):
    onoma = input("Δώσε το ονοματεπώνυμο του μαθητή: ")
    vathmos = int(input("Δώσε το βαθμό: "))
```

Για τον έλεγχο της βαθμολογίας θα πρέπει να χρησιμοποιήσουμε τη δομή `while`.
Η συνθήκη της θέλουμε να γίνεται *αληθής* όταν έχουμε *λάθος δεδομένα*.

```python
for ...:
    onoma = ...
    vathmos = int(input("Δώσε το βαθμό: "))
    while vathmos < 1 or vathmos > 100:  # Όσο ο βαθμός είναι μικρότερος του 1 Ή μεγαλύτερος του 100
        print("Λάθος βαθμός")
        vathmos = int(input("Δώσε το βαθμό: "))  # ΥΠΟΧΡΕΩΤΙΚΑ διαβαζουμε ξανά τον βαθμό!
```

Τώρα θα πρέπει να καταχωρήσουμε τα δύο δεδομένα σε δύο λίστες.

```python
onomata = []
vathmoi = []
for ...:
    onoma = ...
    vathmos...
    while ...:
        print...
        vathmos = ...
    onomata.append(onoma)  # Πρόσθεσε στη λίστα onomata το onoma που διάβασες
    vathmoi.append(vathmos)  # Πρόσθεσε στη λίστα vathmoi το vathmos που διάβασες
```

Η παραπάνω εισαγωγή δεδομών δημιουργεί δύο **παράλληλες** λίστες. Δηλαδή, στη 
θέση 0 της λίστας `onomata` έχουμε ένα όνομα, και στη θέση 0 της λίστας `vathmoi`
τον αντίστοιχο βαθμό του.

````python
onomata = []
vathmoi = []
for i in range(3):
    onoma = input("Δώσε το ονοματεπώνυμο του μαθητή: ")
    vathmos = int(input("Δώσε το βαθμό: "))
    while vathmos < 1 or vathmos > 100:
        print("Λάθος βαθμός")
        vathmos = int(input("Δώσε το βαθμό: "))
    onomata.append(onoma)  # Πρόσθεσε στη λίστα onomata το onoma που διάβασες
    vathmoi.append(vathmos)  # Πρόσθεσε στη λίστα vathmoi το vathmos που διάβασες

print(onomata)
print(vathmoi)
````

# 2ο Ζητούμενο

Όταν χρησιμοποιούμε λίστες, καλό είναι να υπολογίζουμε άθροισμα, μέσο όρο,
μικρότερο/μεγαλύτερο, *αφού έχουμε τελειώσει με την εισαγωγή των δεδομένων*.
Αυτό σημαίνει *μετά* τη `for` ή τη `while` στην οποία γεμίζουμε τις λίστες.

Για να υπολογίσουμε τον μέσο όρο, χρειαζόμαστε μόνο τη λίστα με τις βαθμολογίες.
Δεν χρειαζόμαστε τις θέσεις των βαθμών μέσα στη λίστα, οπότε μπορούμε να 
χρησιμοποιούμε τη *δομή διάσχισης λίστας*.

Για να βρούμε τον μέσο όρο χρειαζόμαστε το **άθροισμα των βαθμών** να το 
διαιρέσουμε με το **πλήθος των βαθμών**.

```python
athroisma = 0  # Άθροισμα βαθμών
for vathmos in vathmoi:
    athroisma = athroisma + vathmos
N = len(vathmoi)  # Πλήθος βαθμών
mesos_oros = athroisma / N
print("Ο μέσος όρος είναι: ", mesos_oros)
```

````python
athroisma = 0  # Άθροισμα βαθμών
for vathmos in vathmoi:
    athroisma = athroisma + vathmos
N = len(vathmoi)  # Πλήθος βαθμών
mesos_oros = athroisma / N
print("Ο μέσος όρος είναι: ", mesos_oros)
````

# 3ο Ζητούμενο

Για αυτό το ζητούμενο θα πρέπει να διασχίσουμε τη λίστα `vathmoi` και να 
συγκρίνουμε τον κάθε βαθμό με τον μέσο όρο. Όταν ένα βαθμός είναι μεγαλύτερος
ή ίσος με το μέσο όρο, θέλουμε να εμφανίσουμε όνομα και βαθμό.

Αυτό μπορούμε να το κάνουμε μόνο με τη *δομή διάσχισης θέσης λίστας*.
Δηλαδή, όταν βρίσκουμε ένα βαθμό στη θέση `i` να είναι `>=mesos_oros`, θα πρέπει
να εμφανίσουμε το `onomata[i]` για το όνομα:

```python
for i in range(len(vathmoi)):  # Για να διασχίσει όλη τη λίστα, στο range βάζουμε len()
    if vathmoi[i] >= mesos_oros:  # Αν ο βαθμός στη θέση i είναι μεγαλύτερος ή ίσος με τον μέσο όρο
        print(onomata[i], vathmoi[i])  # Εμφάνισε το όνομα στη θέση i και το βαθμό στη θέση i
```

````python
for i in range(len(vathmoi)):  # Για να διασχίσει όλη τη λίστα, στο range βάζουμε len()
    if vathmoi[i] >= mesos_oros:  # Αν ο βαθμός στη θέση i είναι μεγαλύτερος ή ίσος με τον μέσο όρο
        print(onomata[i], vathmoi[i])  # Εμφάνισε το όνομα στη θέση i και το βαθμό στη θέση i
````

# Ολοκληρωμένο Πρόγραμμα

````python
onomata = []  # Λίστα για την εισαγωγή των ονομάτων
vathmoi = []  # Λίστα για την εισαγωγή των βαθμών
for i in range(10):  # Έχουμε συνολικά 10 μαθητές
    onoma = input("Δώσε το ονοματεπώνυμο του μαθητή: ")
    vathmos = int(input("Δώσε το βαθμό: "))
    while vathmos < 1 or vathmos > 100:  # Όσο ο βαθμός είναι μικρότερος του 1 Ή μεγαλύτερος του 100, δηλαδή λάθος
        print("Λάθος βαθμός")
        vathmos = int(input("Δώσε το βαθμό: "))  # Διάβασε ξανά τον βαθμό
    onomata.append(onoma)  # Πρόσθεσε στη λίστα onomata το onoma που διάβασες
    vathmoi.append(vathmos)  # Πρόσθεσε στη λίστα vathmoi το vathmos που διάβασες

athroisma = 0  # Άθροισμα βαθμών
for vathmos in vathmoi:  # Πάρε έναν-έναν τους βαθμούς από τη λίστα vathmoi
    athroisma = athroisma + vathmos  # Πρόσθεσε τον βαθμό στο άθροισμα
N = len(vathmoi)  # Πλήθος βαθμών
mesos_oros = athroisma / N  # Υπολόγισε τον μέσο όρο
print("Ο μέσος όρος είναι: ", mesos_oros)

for i in range(len(vathmoi)):  # Για να διασχίσει όλη τη λίστα, στο range βάζουμε len()
    if vathmoi[i] >= mesos_oros:  # Αν ο βαθμός στη θέση i είναι μεγαλύτερος ή ίσος με τον μέσο όρο
        print(onomata[i], vathmoi[i])  # Εμφάνισε το όνομα στη θέση i και το βαθμό στη θέση i
````
