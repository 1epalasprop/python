# Εκφώνηση

Η μέγιστη επιτρεπόμενη τιμή της περιεκτικότητας κατ’ όγκο των καυσαερίων σε 
οξείδια του αζώτου (NOx) ανηγμένη σε οξυγόνο αναφοράς 3%, σε ppm είναι 150.
Σε περίπτωση που ο ρύπος αυτός ξεπεράσει τα 150 ppm πρέπει να ληφθούν μέτρα.
Τα υπουργείο περιβάλλοντος έχει εγκαταστήσει 50 σταθμούς μέτρησης του ρύπου
σε διάφορα σημεία της χώρας.

Να αναπτύξετε πρόγραμμα στην γλώσσα προγραμματισμού Python το οποίο

1. Να διαβάζει την αριθμητική τιμή του ρύπου για κάθε σταθμό μέτρησης.
   Η αριθμητική τιμή αυτή να είναι θετικός αριθμός.
1. Να υπολογίζει και να εμφανίζει το πλήθος των σταθμών με ένδειξη μεγαλύτερη από 150.
1. Να υπολογίζει και να εμφανίζει τη μέση τιμή του ρύπου στους 50 σταθμούς.

## 1ο Ζητούμενο

Το 1ο ζητούμενο έχει δύο (2) μέρη:
1. Να την τιμή των ρύπων **σε κάθε σταθμό**
2. Να ελέγχει ότι η τιμή είναι **θετικός αριθμός**

Για να διαβάσουμε τους ρύπους σε κάθε σταθμό, **αφού γνωρίζουμε πόσους σταθμούς**
έχει, χρησιμοποιούμε τη `for`.

```python
for stathmos in range(50):
    nox = int(input("Δώσε τους ρύπους του σταθμού: "))
```

Για να ελέγξουμε ότι η μεταβλητή `nox` θα έχει θετικό αριθμό, χρησιμοποιούμε
*έλεγχο εγκυρότητας* με τη `while`.

```python
for ...:
    nox = ...
    while nox < 0:  # Όσο οι ρύποι είναι αρνητικοί
        print("Λάθος τιμή")  # Προαιρετικό αλλά χρήσιμο
        nox = int(input("Δώσε τους ρύπους του σταθμού: "))
```

## 2ο Ζητούμενο

Στο ζητούμενο λέει "**το πλήθος των σταθμών**. Αυτό σημαίνει ότι χρειαζόμαστε
μία μεταβλητή που θα αυξάνει ότι η τιμή του `nox` μέσα στη `for` γίνει μεγαλύτερη
από 150.

```python
plithos = 0  # Το πλήθος των σταθμών με nox > 150
for ...:
    nox = ...
    while ...:
        print...
        nox = ...
    # Αυξάνω το πλήθος μόνο αν nox > 150
    if nox > 150:
        plithos = plithos + 1  # Πρόσθεσε ένα σταθμό
```

## 3ο Ζητούμενο

Για να υπολογίσουμε τον μέσο όρο, χρειάζεται να υπολογίσουμε το **άθροισμα όλων των ρύπων**.
Χρειαζόμαστε μία μεταβλητή γι' αυτό το σκοπό και θα πρέπει να πάρει αρχική τιμή **πρίν**
από τη `for` γιατί **αφορά όλους τους σταθμούς**.

Τον μέσο όρο μπορούμε να τον υπολογίζουμε **μόνο όταν έχουμε ολοκληρώσει τη `for`**.
Αυτό σημαίνει ότι ο υπολογισμός και η εμφάνιση θα γίνει **μετά** τη `for`.

Για τον υπολογισμό του μέσου όρου χρειαζόμαστε δύο όρους: *α)* τον αριθμητή,
που είναι το **άθροισμα**, και *β)* τον παρανομαστή που είναι το **πλήθος**
των επαναλήψεων που έχουμε κάνει. Το άθροισμα το υπολογίζουμε μέσα στη `for`.
Το πλήθος το ξέρουμε, αφού έχουμε 50 σταθμούς.

```python
athroisma = 0  # Συνολικό άθροισμα πόντων
plithos = ...
for ...:
    nox = ...
    while ...:
        print...
        nox = ...
    if ...:
        plithos = ...
    # Πρόσθεσε τους ρύπους του σταθμού στο άθροισμα
    athroisma = athroisma + nox
# Έξω από τη for
mesos_oros = float(athroisma) / 50  # 50 είναι οι επαναλήψεις του for
print("Ο μέσος όρος των ρύπων είναι: ", mesos_oros)
```

# Ολοκληρωμένο Πρόγραμμα

````python
athroisma = 0  
plithos = 0  
for stathmos in range(1, 51):
    nox = int(input("Δώσε τους ρύπους του σταθμού: "))
    while nox < 0:
        print("Λάθος τιμή")
        nox = int(input("Δώσε τους ρύπους του σταθμού: "))
    if nox > 150
        plithos = plithos + 1
    athroisma = athroisma + nox
mesos_oros = float(athroisma) / 50
print("Ο μέσος όρος των ρύπων είναι: ", mesos_oros)
````
