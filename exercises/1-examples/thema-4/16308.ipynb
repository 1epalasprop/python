# Εκφώνηση

Σε ένα εργοστάσιο κατασκευής τρισδιάστατων εκτυπωτών απασχολούνται 35 υπάλληλοι.
Ο μηνιαίος μισθός του κάθε υπαλλήλου προκύπτει με τον παρακάτω τρόπο:

* ο κάθε υπάλληλος έχει βασικό μισθό 720 €,
* για κάθε ώρα υπερωρίας που εργάστηκε μέσα στον μήνα αμείβεται με 5 € ανά ώρα,
* για κάθε τρία χρόνια εργασίας λαμβάνει χρονοεπίδομα 25 € ανά τριετία, π.χ.,
  εάν κάποιος υπάλληλος έχει 10 χρόνια εργασίας, θα λάβει 3 * 25=75 € επίδομα,
* το άθροισμα όλων των παραπάνω δημιουργούν τον **ακαθάριστο μισθό** του
  υπαλλήλου.

Στον ακαθάριστο μισθό του κάθε υπαλλήλου, που υπολογίζεται όπως παραπάνω, υπάρχουν
οι παρακάτω κρατήσεις:

* 8% για το ταμείο της ασφάλισης του υπαλλήλου, και
* το 7% για την σύνταξη του υπαλλήλου.

Να αναπτύξετε πρόγραμμα στην γλώσσα προγραμματισμού Python το οποίο

1. Να διαβάζει για κάθε υπάλληλο το ονοματεπώνυμο, τον αριθμό υπερωριών που
   εργάστηκε κατά τον μήνα Μάρτιο καθώς και τα χρόνια εργασίας του υπαλλήλου.
   Τα χρόνια εργασίας να είναι μη αρνητικός αριθμός.
1. Να υπολογίζει τον ακαθάριστο μισθό του υπαλλήλου (όπως περιγράφεται παραπάνω)
   και θα τον εμφανίζει μετά το μήνυμα «Ο μικτός σου μισθός είναι».
1. Να υπολογίζει και να εμφανίζει το ονοματεπώνυμο του υπαλλήλου και τις καθαρές
   αποδοχές του μετά το μήνυμα «Οι καθαρές σου αποδοχές είναι».
1. Να υπολογίζει και να εμφανίζει το σύνολο των κρατήσεων από όλους τους υπαλλήλους
   μετά το μήνυμα «Οι συνολικές κρατήσεις των 35 υπαλλήλων είναι».

## 1ο Ζητούμενο

Το 1ο ζητούμενο έχει δύο (2) μέρη:
1. Να διαβάζει ονοματεπώνυμο και αριθμό υπερωριών και τα χρόνια **για κάθε υπάλληλο**
2. Να ελέγχει ότι ο αριθμός υπερωριών είναι **θετικός αριθμός**

Για να διαβάσουμε τις τιμές, **αφού γνωρίζουμε για πόσους υπαλλήλους**,
χρησιμοποιούμε τη `for`.

```python
for i in range(35):
    name = intpu("Δώσε το όνομα του υπαλλήλου: ")
    over_time = int(input("Δώσε τις υπερωρίες: "))
    years = int(input("Πόσα χρόνια έχει ο υπάλληλος; "))
```

Για να ελέγξουμε ότι η μεταβλητή `years` θα έχει θετικό αριθμό, χρησιμοποιούμε
*έλεγχο εγκυρότητας* με τη `while`.

```python
for ...:
    name = ...
    over_time = ...
    while years < 0:  # Όσο οι ρύποι είναι αρνητικοί
        print("Λάθος τιμή")  # Προαιρετικό αλλά χρήσιμο
        years = int(input("Πόσα χρόνια έχει ο υπάλληλος; "))
```

## 2ο Ζητούμενο

*Για κάθε υπάλληλο* θα πρέπει στον βασικό μισθό του να υπολογίσουμε τον
ακαθάριστο μισθό του. Δηλαδή, ο υπολογισμός γίνεται **μέσα στη** `for`.

```python
for ...:
    name = ...
    over_time = ...
    years = ...
    while ...:
        print...
        years = ...
    basikos = 720  # Ο βασικός μισθός
```

Στον βασικό μισθό θα πρέπει να προσθέσουμε τις υπερωρίες (`over_time`) επί 5€,
αφού για κάθε ώρα παίρνει 5€.

```python
for ...:
    name = ...
    over_time = ...
    years = ...
    while ...:
        print...
        years = ...
    basikos = 720  # Ο βασικός μισθός
    over_salary = over_time * 5  # Μισθός από υπερωρίες
```

Για να προσθέσουμε και τις **τριετίες**, θα πρέπει να τις υπολογίσουμε.
Αρκεί να διαιρέσουμε τα χρόνια εργασίας με το 3.

Παραδείγματα υπολογισμού:

| Χρόνια / 3 | Τριετίες |
|:----------:|:--------:|
| 1/3        | 0        |
| 3/3        | 1        |
| 5/3        | 1        |
| 7/3        | 2        |
| 9/3        | 3        |
| 10/3       | 3        |

Άρα, θα διαιρέσουμε με το τρία και θα πάρουμε το ακέραιο μέρος. Για το 
μισθό από τριετίες θα πρέπει να πολλαπλασιάσουμε επί 25 γιατί δίνει
25€ ανά τριετία.

```python
for ...:
    name = ...
    over_time = ...
    years = ...
    while ...:
        print...
        years = ...
    basikos = 720  # Ο βασικός μισθός
    over_salary = over_time * 5  # Μισθός από υπερωρίες
    trieties = int(years / 3)  # Τριετίες
    trieties_salary = trieties * 25  # Μισθός από τριετίες
```

Τέλος, για να υπολογίσουμε τον **ακαθάριστο μισθό** θα πρέπει να τα αθροίσουμε όλα.

```python
for ...:
    name = ...
    over_time = ...
    years = ...
    while ...:
        print...
        years = ...
    basikos = 720  # Ο βασικός μισθός
    over_salary = over_time * 5  # Μισθός από υπερωρίες
    trieties = int(years / 3)  # Τριετίες
    trieties_salary = trieties * 25  # Μισθός από τριετίες
    akatharistos = basikos + over_salary + trieties_salary  # Ακαθάριστος μισθός
    print("Ο μικτός μισθός είναι:", akatharistos)
```

## 3o Ζητούμενο

Για να υπολογίσουμε τον καθαρό μισθό, θα πρέπει από τον ακαθάριστο να
αφαιρέσουμε τις κρατήσεις. Οι κρατήσεις είναι δύο, ασφάλισης και σύνταξης.
Αφού τις υπολογίσουμε, θα τις αφαιρέσουμε από τον ακαθάριστο μισθό. Το
αποτέλεσμα είναι ο *καθαρός* μισθός.

```python
for ...:
    name = ...
    over_time = ...
    years = ...
    while ...:
        print...
        years = ...
    basikos = ...
    over_salary = ...
    trieties = ...
    trieties_salary = ...
    akatharistos = ...
    print...
    asfalisi = akatharistos * 8 / 100  # 8%
    syntaxi = akatharistos * 7 / 100  # 7%
    kratiseis = asfalisi + syntaxi  # Οι συνολικές κρατήσεις
    katharos = akatharistos - kratiseis  # ακαθάριστος - κρατήσεις
    print("Ο υπάλληλος", name, "παίρνει καθαρό μισθό", katharos, "€")
```

## 4ο Ζητούμενο

Για να υπολογίσουμε τις **συνολικές** κρατήσεις χρειαζόμαστε ένα *άθροισμα*
για *όλους* τους υπαλλήλους. Αυτό σημαίνει ότι θέλουμε μία μεταβλητή που θα
πάρει τιμή **πριν** από το `for` και θα αυξάνει **μέσα** στη `for`.
Το τελικό μήνυμα θα είναι **έξω** από τη `for` γιατί τότε μόνο ξέρουμε το 
τελικό σύνολο κρατήσεων.

```python
ahtroisma_kratisewn = 0  # Άθροισμα κρατήσεων
for ...:
    name = ...
    over_time = ...
    years = ...
    while ...:
        print...
        years = ...
    basikos = ...
    over_salary = ...
    trieties = ...
    trieties_salary = ...
    akatharistos = ...
    print...
    asfalisi = ...
    syntaxi = ...
    kratiseis = ...
    katharos = ...
    print...
    athroisma_kratisewn = athroisma_kratisewn + kratiseis  # Πρόσθεσε στο άθροισμα τις κρατήσεις του υπαλλήλου
# Έξω από τη for
print("Οι συνολικές κρατήσεις των 35 υπαλλήλων είναι:", athroisma_kratisewn, "€")
```

# Ολοκληρωμένο Πρόγραμμα

````python
ahtroisma_kratisewn = 0 
for i in range(35):
    name = input("Δώσε το όνομα του υπαλλήλου: ")
    over_time = int(input("Δώσε τις υπερωρίες: "))
    years = int(input("Πόσα χρόνια έχει ο υπάλληλος; "))
    while years < 0:
        print("Λάθος τιμή")
        years = int(input("Πόσα χρόνια έχει ο υπάλληλος; "))
    basikos = 720
    over_salary = over_time * 5
    trieties = int(years / 3)
    trieties_salary = trieties * 25
    akatharistos = basikos + over_salary + trieties_salary
    print("Ο ακαθάριστος μισθός είναι:", akatharistos, "€")
    asfalisi = akatharistos * 8 / 100
    syntaxi = akatharistos * 7 / 100
    kratiseis = asfalisi + syntaxi
    katharos = akatharistos - kratiseis
    print("Ο υπάλληλος", name, "παίρνει καθαρό μισθό", katharos, "€")
    athroisma_kratisewn = athroisma_kratisewn + kratiseis
print("Οι συνολικές κρατήσεις των 35 υπαλλήλων είναι:", athroisma_kratisewn, "€")
````
