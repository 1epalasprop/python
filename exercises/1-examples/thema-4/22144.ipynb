# Εκφώνηση

Σε μία κινηματογραφική ταινία σε σινεμά το εισιτήριο εισόδου είναι για τους ενήλικες επτά (7) ευρώ και για τα παιδιά τέσσερα (4) ευρώ. Η χωρητικότητα του σινεμά είναι διακόσια (200) άτομα. Να αναπτύξετε πρόγραμμα σε γλώσσα προγραμματισμού Python, το οποίο να υλοποιεί τα παρακάτω:

1. Για κάθε άτομο να διαβάζει την τιμή "E" για ενήλικας ή "Ρ" για παιδί. Η διαδικασία της εισόδου θεατών να τερματίζεται, όταν εισαχθεί η τιμή "TELOS". Δεν απαιτείται έλεγχος εγκυρότητας εισαγωγής δεδομένων.
2. Να υπολογίζει και να εμφανίζει στην οθόνη τα συνολικά έσοδα του κινηματογράφου και το πλήθος των κενών θέσεων.
3. Τέλος, να υπολογίζει και να εμφανίζει το ποσοστό των παιδιών στο σύνολο των θεατών. Θεωρήστε ότι υπάρχει τουλάχιστον ένας θεατής-παιδί.

# Λύση

Από την εκφώνηση έχουμε ότι το εισητήριο των ενηλίκων είναι €7, των παιδιών €4, και η χωρητικότητα του σινεμά 200 άτομα.
Αυτά τα δίνουμε ως τιμές σε μεταβλητές.

```python
adult_price = 7  # Εισητήριο ενηλίκων
child_price = 4  # Εισητήριο παιδιών
seats = 200  # Θέσεις
```

## Ζητούμενο 1
Μας ζητείται να διαβάζουμε επαναληπτικά το άτομο μέχρι να εισαχθεί η τιμή "TELOS".
Αυτό σημαίνει ότι θέλουμε μια δομή *`while`*, και συγκεκριμένα το μοτίβο:
```python
person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε TELOS για τερματισμό")
while person != "TELOS":
    person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε TELOS για τερματισμό")
```
**ΕΠΙΣΗΜΑΝΣΗ:** Δεν ζητείται έλεγχος στην τιμή που δώσαμε στη `person`. Δηλαδή, 
αν αυτή είναι "Ε" ή "Ρ" και μόνο.

Υπάρχει όμως μία ακόμη συνθήκη, η χωρητικότητα της αίθουσας. Δεν μπορούμε, δηλαδή,
να έχουμε πάνω από 200 θεατές. Έτσι, θα πρέπει να το βάλουμε στη συνθήκη της while.
Θα χρειαστούμε όμως έναν μετρητή για να ξέρουμε πόσοι θεατές εχουν μπει. Έτσι,
ο κώδικας γίνεται:

```python
count = 0  # Πόσοι είναι στην αίθουσα
person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε TELOS για τερματισμό")
while person != "TELOS" and count < 200:  # Όσο δε δώθηκε "TELOS" και μέσα στην αίθουσα είναι κάτω από 200
    count = count + 1  # Ένα άτομο μπήκε. Αύξησε τον μετρητή.
    person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε TELOS για τερματισμό")
```

## Ζητούμενο 2

Για να υπολογίσουμε τα συνολικά έσοδα χρειαζόμαστε μία μεταβλητή (έστω `esoda`)
όπου θα τα προσθέτουμε. Αυτή πρέπει να μπει πριν τη `while` γιατί αφορά *όλους*
τους θεατές.

Επίσης, *μέσα* στη `while` θα πρέπει να ελέγχουμε τη μεταβλητή `person` και 
ανάλογα την τιμή της (Ε ή Ρ) θα προσθέτουμε στα `esoda` το `adult_price` ή
το `child_price` αντίστοιχα.

```python
esoda = 0  # Ξεκινάμε με 0 έσοδα
count = 0  # Πόσοι είναι στην αίθουσα
person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε TELOS για τερματισμό")
while person != "TELOS" and count < 200:  # Όσο δε δώθηκε "TELOS" και μέσα στην αίθουσα είναι κάτω από 200
    count = count + 1  # Ένα άτομο μπήκε. Αύξησε τον μετρητή.
    # Ενημέρωσε τα έσοδα
    if person == "E":
        esoda = esoda + adult_price
    elif person == "P":
        esoda = esoda + child_price
    person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε TELOS για τερματισμό")
```

Θα εμφανίσουμε τα συνολικά έσοδα και τον αριθμό των κενών θέσεων έξω από τη
`while`.

```python
print("Τα συνολικά έσοδα είναι:", esoda)
```

Για να υπολογίσουμε τις κενές θέσεις, αρκεί από τη μεταβλητή `seats` να 
αφαιρέσουμε τη μεταβλητή `count`.

```python
empty = seats - count
print("Το πλήθος των κενών θέσεων είναι: ")
```

## Ζητούμενο 3

Για να υπολογίσουμε το ποσοστό των παιδιών θα πρέπει να ξέρουμε πόσα είναι.
Χρειαζόμαστε, λοιπόν, έναν μετρητή πριν τη `while`. Επίσης, θα πρέπει να
αυξάνει μόνο όταν το `person == "P"`.

Άρα έχουμε:
```python
count_child = 0  # Μετρητής παιδιών
while ...:
    ...
    elif person == "P":
        ...
        count_child = count_child + 1  # Αύξηση των παιδιών
```

Για να υπολογίσουμε το ποσοστό των παιδιών θα πρέπει να κάνουμε την πράξη:
```
πλήθος παιδιών / πλήθος θεατών * 100
```
Έχουμε και τις δύο μεταβλητές που χρειαζόμαστε, έτσι γράφουμε:
```python
childs = count_child / count * 100
```

# Ολοκληρωμένο Πρόγραμμα

````python
    adult_price = 7  # Εισητήριο ενηλίκων
child_price = 4  # Εισητήριο παιδιών
seats = 20  # Θέσεις (έβαλα 20 για να μπορούμε να το τρέξουμε εύκολα)
esoda = 0  # Ξεκινάμε με 0 έσοδα
count = 0  # Πόσοι είναι στην αίθουσα
count_child = 0  # Παιδιά στην αίθουσα
person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε T για τερματισμό")
while person != "T" and count < 200:  # Όσο δε δώθηκε "TELOS" και μέσα στην αίθουσα είναι κάτω από 200
    count = count + 1  # Ένα άτομο μπήκε. Αύξησε τον μετρητή.
    # Ενημέρωσε τα έσοδα
    if person == "E":  # Αν το άτομο είναι ενήλικας
        esoda = esoda + adult_price  # Πρόσθεσε στα έσοδα το εισητήριο των ενηλίκων
    elif person == "P":  # Αλλιώς ανε είναι παιδί
        esoda = esoda + child_price  # Πρόσθεσε στα έσοδα το εισητήριο των παιδιών
        count_child = count_child + 1  # Μέτρα ότι ένα παιδί μπήκε στην αίθουσα
    # Ρώτα για τον επόμενο θεατή.
    person = input("Ενήλικας (Ε) ή παιδί (Ρ); Δώστε T για τερματισμό")

# Εμφάνιση αποτελεσμάτων
print("Τα συνολικά έσοδα είναι: €", esoda)
empty = seats - count
print("Το πλήθος των κενών θέσεων είναι:", empty)
childs = count_child / count * 100
print("Το ποσοστό των παιδιών στην αίθουσα είναι:", childs, "%")
````
