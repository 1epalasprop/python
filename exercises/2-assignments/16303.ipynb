# Εκφώνηση

Στο ακριτικό χωριό Σύνορα έχουν απογραφεί, με την τελευταία απογραφή, 380 άτομα.
Να γραφεί πρόγραμμα σε γλώσσα Python, που θα χρησιμοποιηθεί για την νέα απογραφή
έτσι ώστε να εξαχθούν κάποια απαραίτητα στατιστικά, το οποίο:

* Να διαβάζει **για κάθε** κάτοικο την ηλικία του μέχρι να δοθεί αρνητικός αριθμός
  ή 0. Δεν χρειάζεται έλεγχος έγκυρης τιμής. Θεωρείστε ότι στην νέα απογραφή
  δίνεται τουλάχιστον μια ηλικία.
* Να υπολογίζει πόσα άτομα απογράφηκαν και αν υπάρχει μείωση ή αύξηση από την
  προηγούμενη απογραφή καθώς και πόσο ήταν αυτή, μετά από κατάλληλο μήνυμα.
* Να υπολογίζει και να εμφανίζει τον μέσο όρο ηλικίας των κατοίκων του χωριού, με την
  νέα απογραφή.
* Να υπολογίζει και να εμφανίζει την μικρότερη και την μεγαλύτερη ηλικία που έχει
  απογραφεί στην νέα απογραφή.
