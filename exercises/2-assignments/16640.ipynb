# Εκφώνηση

Μια αεροπορική εταιρεία πρόκειται να προσφέρει ειδικό επίδομα (bonus) στους 300
ιπτάμενους φροντιστές (αεροσυνοδούς) που απασχολεί. Το bonus υπολογίζεται σαν
ένα ποσοστό των ετήσιων αποδοχών τους, ανάλογα με τα μίλια πτήσεων που
πραγματοποίησαν συνολικά κατά το περασμένο έτος, με βάση τον παρακάτω πίνακα:

| Ετήσια μίλια πτήσεων | Ποσοστό bonus (επί των ετήσιων αποδοχών) |
|:--------------------:|:----------------------------------------:|
| 200000 - 250000      | 5%                                       |
| 250001 - 300000      | 10%                                      |
| > 300000             | 15%                                      |

Οι ιπτάμενοι φροντιστές που πραγματοποίησαν λιγότερα από 200000 μίλια πτήσεων
ετησίως, δε δικαιούνται bonus. Ο υπολογισμός του bonus *δεν γίνεται κλιμακωτά*. Να γραφεί πρόγραμμα σε Python που:

1. Για κάθε ιπτάμενο φροντιστή διαβάζει τον κωδικό εργαζομένου του (EMPL), τα
   μίλια πτήσεων (MIL) που συμπλήρωσε κατά το περασμένο έτος και τις ετήσιες
   αποδοχές του σε ευρώ (SAL), χωρίς να πραγματοποιείται έλεγχος ορθότητας των
   πληκτρολογούμενων δεδομένων.
2. Υπολογίζει το ποσό του bonus που δικαιούται ο συγκεκριμένος ιπτάμενος
   φροντιστής, με βάση τον παραπάνω πίνακα και το εμφανίζει, εφόσον δεν είναι
   μηδενικό. Αν ο ιπτάμενος φροντιστής δε δικαιούται bonus, εμφανίζει το μήνυμα
   **“Ο ιπτάμενος φροντιστής δε δικαιούται bonus!”**.3. Βρίσκει κι εμφανίζει τον κωδικό εργαζομένου του φροντιστή αέρος με τα
   περισσότερα ετήσια μίλια πτήσης, καθώς και πόσα ήταν τα ετήσια μίλια του.
   Θεωρούμε πως είναι μοναδικός.
