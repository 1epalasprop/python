# Εκφώνηση

Το Μουσείο Φυσικής Ιστορίας της περιοχής σας προσφέρει 5 εκπαιδευτικά προγράμματα για παιδιά. Το εισιτήριο για κάθε παιδί πωλείται προς 7 ευρώ. Το Μουσείο καταγράφει τα εισιτήρια που πουλήθηκαν τον περασμένο μήνα ανά πρόγραμμα, με σκοπό να αποφασίσει τη συνέχιση ή μη του κάθε προγράμματος. Κριτήριο για τη συνέχιση κάθε προγράμματος είναι να έχει φέρει έσοδα στο Μουσείο τουλάχιστον 1500 ευρώ κατά τον περασμένο μήνα. 
 Να γραφεί πρόγραμμα σε Python πο:

1. Για κάθε εκπαιδευτικό πρόγραμμα διαβάζει τον τίτλο του προγράμματος. Διαβάζει, επίσης το πλήθος των εισιτηρίων που πουλήθηκαν για το πρόγραμμα αυτό τον περασμένο μήνα, ελέγχοντας ότι ο αριθμός που καταχωρείται είναι μεγαλύτερος ή ίσος του μηδενός. (Μονάδες 6)
2. Το πρόγραμμα υπολογίζει τα μηνιαία έσοδα από τις πωλήσεις εισιτηρίων του προγράμματος αυτού. Εμφανίζει, ακολούθως, τον τίτλο του προγράμματος, τα μηνιαία έσοδά του και κατάλληλο μήνυμα για τη συνέχιση ή μη του προγράμματος. (Μονάδες 7)
3. Υπολογίζει κι εμφανίζει το πλήθος των προγραμμάτων που πούλησαν περισσότερα από 1000 εισιτήρια το καθένα. (Μονάδες 7)
4. Υπολογίζει κι εμφανίζει το συνολικό ποσό σε ευρώ που εισέπραξε το Μουσείο τον περασμένο μήνα από τις πωλήσεις εισιτηρίων και των 5 εκπαιδευτικών προγραμμάτων. (Μονάδες 5)υ
