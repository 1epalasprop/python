# Εκφώνηση 

Κατά την Παγκόσμια ημέρα της Αποταμίευσης, ένας καθηγητής Πληροφορικής, θέλοντας να κεντρίσει το ενδιαφέρον των μαθητών του, τους έβαλε το ακόλουθο κουίζ: 

Σε πόσες ημέρες θα καταφέρουν να αποταμιεύσουν το ποσό των 100€ αν ξεκινήσουν την 1η ημέρα από το αρχικό ποσό του ενός λεπτού του ευρώ (0.01 €) και κάθε επόμενη μέρα προστίθεται (αποταμιεύεται) στον κουμπαρά ποσό διπλάσιο από την προηγούμενη (την 1η ημέρα 0.01€, τη 2η ημέρα 0.02€, την 3η μέρα 0.04€, κ.ο.κ). Για να προσομοιώσουν την διαδικασία της αποταμίευσης, ζήτησε από τους μαθητές να γράψουν ένα πρόγραμμα σε Python το οποίο: 
1. Για κάθε μέρα να εμφανίζει το ποσό που έχει ήδη αποταμιευθεί στο κουμπαρά (π.χ. για την 1η ημέρα θα εμφανίζει το μήνυμα «Στο κουμπαρά έχει αποταμιευθεί το ποσό των 0.01€»), μέχρι να συγκεντρωθεί το συνολικό ποσό των 100€. (Μονάδες 10)
2. Να εμφανίζει πόσες μέρες χρειάστηκαν για ν’ αποταμιευθεί το ποσό των 100€. (Μονάδες 5)
3. Να υπολογίζει κι εμφανίζει κατά πόσο ξεπεράστηκε το ποσό των 100€. (Μονάδες 5)
4. Τέλος, το πρόγραμμα θα πρέπει να υπολογίζει κι εμφανίζει ποιο ήταν το ποσό που τοποθετήθηκε στον κουμπαρά την τελευταία μέρα. (Μονάδες 5)
