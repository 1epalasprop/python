# Εύρεση του Μεγαλύτερου/Μικρότερου Στοιχείου

Αν μας ενδιαφέρει μόνο να βρούμε το μεγαλύτερο ή το μικρότερο στοιχείο *χωρίς*
να θέλουμε την *θέση* του στη λίστα, τότε θα χρησιμοποιήσουμε *διάσχιση
στοιχείων λίστας*, διαφορετικά *διάσχιση θέσεων λίστας*.

```
                       +---ΟΧΙ---> Διάσχιση στοιχείων λίστας (for item in list)
                       |
Θέλω να ξέρω τη θέση του μεγαλύτερου/μικρότερου;
                       |
                       +---ΝΑΙ---> Διάσχιση θέσεων λίστας (for i in range(n))
```

**ΠΑΡΑΤΗΡΗΣΗ:** Η ίδια διαδικασία εφαρμόζεται σε λίστες αριθμό και συμβολοσειρών.
Δηλαδή, αν η λίστα μου έχει *μόνο αριθμούς* ή *μόνο συμβολοσειρές* ο **ίδιος**
κώδικας θα δουλέψει.

## Εύρεση με Διάσχιση Στοιχείων Λίστας

Ακολουθούμε την εξής διαδικασία για να βρούμε το *μεγαλύτερο*:

1. Θέτουμε σε μία μεταβλητή (`maxi`) την τιμή του στοιχείου στη *θέση 0*.
   Αυτή θεωρούμε ότι μέχρι τώρα είναι η μεγαλύτερη στη λίστα.
1. Χρησιμοποιούμε τη δομή `for item in list` για να πάρουμε στην `item`
   διαδοχικά ένα-ένα τα στοιχεία της λίστας.
1. Αν το `item` είναι μεγαλύτερο από το `maxi`, τότε κάνουμε το `maxi` ίσο
   με το `item`
1. Μετά τη `for`, στη `maxi` θα έχουμε το μεγαλύτερο στοιχείο.

Για να βρούμε το *μικρότερο* στοιχείο, το μόνο που αλλάζει είναι ότι στο βήμα
4 χρησιμοποιούμε το μικρότερο αντί του μεγαλύτερου.

### Παράδειγμα
1. Θέτουμε σε μία μεταβλητή (`maxi`) την τιμή του στοιχείου στη *θέση 0*.
   Αυτή θεωρούμε ότι μέχρι τώρα είναι η μεγαλύτερη στη λίστα.
2. Θέτουμε σε μία μεταβλητή (`n`) το πλήθος των στοιχείων της λίστας
   χρησιμοποιώντας τη συνάρτηση `len()`
3. Με τη χρήση 

````python
numbers = [5, -1, 7, 3, 6]
maxi = numbers[0]  # 1
for number in numbers:  # 2
    if number > maxi:  # 3
        maxi = number  # 4
print(maxi)
````

## Εύρεση με Διάσχιση Θέσεων Λίστας

Ακολουθούμε την εξής διαδικασία για να βρούμε το *μεγαλύτερο*:

1. Θέτουμε σε μία μεταβλητή (`maxi`) την τιμή του στοιχείου στη *θέση 0*.
   Αυτή θεωρούμε ότι μέχρι τώρα είναι η μεγαλύτερη στη λίστα.
1. Θέτουμε σε μία μεταβλητή (`n`) το πλήθος των στοιχείων της λίστας
   χρησιμοποιώντας τη συνάρτηση `len()`
1. Χρησιμοποιούμε τη δομή `for i in range(n)` για να πάρουμε διαδοχικά
   μία-μία τις θέσεις των στοιχείων της λίστας.
1. Αν το `my_list[i]` είναι μεγαλύτερο από το `maxi`, τότε κάνουμε το `maxi`
   ίσο με το `my_list[i]`.
1. Μετά τη `for`, στη `maxi` θα έχουμε το μεγαλύτερο στοιχείο.

Για να βρούμε το *μικρότερο* στοιχείο, το μόνο που αλλάζει είναι ότι στο βήμα
5 χρησιμοποιούμε το μικρότερο αντί του μεγαλύτερου.

### Παράδειγμα

````python
numbers = [5, -1, 7, 3, 6]
maxi = numbers[0]  # 1
n = len(numbers)  # 2
for i in range(n):  # 3
    if numbers[i] > maxi:  # 4
        maxi = numbers[i]  # 5
print(maxi)
````
