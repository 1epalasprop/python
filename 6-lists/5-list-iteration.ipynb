# Διάσχιση Λίστας

Με τον όρο *διάσχιση* εννωούμε ότι διαβάζουμε ένα-ένα τα στοιχεία της λίστα
ξεκινώντας από τη θέση 0 και τελειώνοντας στο τελευταίο στοιχείο της.

Αυτή η διαδικασία είναι χρήσιμη για τις ακόλουθες περιπτώσεις:
* Εμφάνιση στοιχείων λίστας
* Υπολογισμός αθροίσματος όταν τα στοιχεία της λίστας είναι αριθμοί
* Υπολογισμός του μεγαλύτερου/μικρότερου στοιχείου της λίστας
* Αναζήτηση ενός στοιχείου της λίστας (υπάρχει/δεν υπάρχει)

Έχουμε δύο τρόπους να διασχίσουμε μία λίστα.
* Έχοντας πρόσβαση στα στοιχεία της λίστας ένα-ένα
* Έχοντας πρόσβαση στη θέση των στοιχείων

# Διάσχιση Στοιχείων Λίστας

Αυτό ο τρόπος είναι χρήσιμος όταν *δεν* μας ενδιαφέρει η θέση των στοιχείων.
Ο τρόπος για να διασχίσουμε τη λίστα είναι:

```python
a_list = ["Αθήνα", "Θεσσαλονίκη", "Πάτρα", "Ηράκλειο"]
for item in a_list:
    print(item)
```

Η μεταβλητή `item` που βρίσκεται στη `for` παίρνει διαδοχικά ένα-ένα τα
στοιχεία της λίστας. Στη συνέχεια, η `print(item)` εμφανίζει το στοιχείο
που βρίσκεται στην item.

Η λίστα *δεν αλλάζει*.

## Παράδειγμα

````python
a_list = ["Αθήνα", "Θεσσαλονίκη", "Πάτρα", "Ηράκλειο"]
for item in a_list:
    print(item)

print("Η λίστα δεν έχει αλλάξει:", a_list)
````

# Διάσχιση Θέσεων Λίστας με for

Αυτός ο τρόπος είναι χρήσιμος όταν *μας* ενδιαφέρει η θέση των στοιχείων.
Ο τρόπος για να διασχίσουμε τη λίστα είναι:

```python
a_list = ["Αθήνα", "Θεσσαλονίκη", "Πάτρα", "Ηράκλειο"]
for i in range(len(a_list)):
    print(a_list[i])
```

**Προσέξτε** τη χρήση της `len()` για να πάρουμε το πλήθος των στοιχείων της
λίστας και της `range()` για να πάρουμε διαδοχικά της θέσεις από το 0.

Στο παραπάνω παράδειγμα έχουμε:
```python
print(len(a_list))
4
```

Δηλαδή, η λίστα έχει **4** στοιχεία. Αυτό το `len(a_list)` βρίσκεται μέσα
στη `range()`, άρα γίνεται `range(4)`, και η `for` έχει γίνει:
```python
for i in range(4):
    print(a_list[i])
```
που θα μας δώσει στο `i` διαδοχικά τις τιμές `0, 1, 2, 3`.

Χρησιμοποιώντας στο `i` ως δείκτη στη λίστα, είναι σαν να έχουμε γράψει:
```python
print(a_list[0])
print(a_list[1])
print(a_list[2])
print(a_list[3])
```

Αντί, λοιπόν, να γράφουμε με το χέρι μία-μία τις θέσεις τη λίστας (που μπορεί
να είναι εκατοντάδες), χρησιμοποιούμε τη `for`.

## Παράδειγμα

````python
a_list = ["Αθήνα", "Θεσσαλονίκη", "Πάτρα", "Ηράκλειο"]
for i in range(len(a_list)):
    print(a_list[i])
````

# Διάσχιση Θέσεων Λίστας με while

Αν θέλουμε (ή πρέπει) να διασχίσουμε τη λίστα με `while` αντί της `for`, τότε
θα πρέπει να το κάνουμε με τις θέσεις των στοιχείων. Ουσιαστικά, αυτό που
κάνουμε είναι μια μετατροπή της `for` σε `while`.

## Παράδειγμα

````python
a_list = ["Αθήνα", "Θεσσαλονίκη", "Πάτρα", "Ηράκλειο"]
i = 0  # Δείκτης θέσης
while i < len(a_list):  # Χρειάζεται η len() για να ξέρει που θα σταματήσει
    print(a_list[i])
    i = i + 1  # Κάνουμε εμείς την αύξηση
````
