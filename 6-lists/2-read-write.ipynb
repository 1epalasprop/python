# Προσπέλαση Στοιχείων Λίστας

Μπορούμε να έχουμε πρόσβαση, για ανάγνωση ή εγγραφή, σε όποιο στοιχείο της
λίστας θέλουμε χρησιμοποιώντας δείκτες. Στην Python οι δείκτες ξεκινούν από
το 0. Στις λίστες, όπως και στις συμβολοσειρές, υπάρχει και ο δείκτης με τις
αρνητικές τιμές.

Χρησιμοποιούμε τον δείκτη με τις θετικές τιμές, όταν θέλουμε να προσπελάσουμε
κάποιο στοιχείο με βάση την αρχή της λίστας, π.χ., το 5ο στοιχείο της λίστα
(που είναι στη θέση 4 γιατί ξεκινάμε να μετράμε από το 0).
Αντίστοιχα, τον δείκτη με τις αρνητικές τιμές τον χρησιμοποιούμε όταν θέλουμε
να προσπελάσουμε κάποιο στοιχείο με βάση το *τέλος* της λίστας, π.χ., το τρίτο
στοιχείο από το τέλος (θέση -3).

```python
Θετικός δείκτης (θέση): 0          1         2         3         4            5          6
        days_of_week = ["Δευτέρα", "Τρίτη", "Τετάρτη", "Πέμπτη", "Παρασκευή", "Σαββάτο", "Κυριακή"]
Αρνητικός δείκτης:      -7         -6       -5         -4        -3           -2         -1
```

## Ανάγνωση
Για να προσπελάσω ένα στοιχείο της λίστας, γράφω το όνομά της, αγκύλες (`[]`)
και τη θέση.

```python
print(days_of_week[0])  # Εμφανίζει: Δευτέρα
print(days_of_week[-2]) # Εμφανίζει: Σαββάτο
```

Οι τιμές μίας λίστας, αν είναι απλοί τύποι δεδομένων (δηλαδή αριθμοί,
συμβολοσειρές και λογικές τιμές) μπορούν να συμμετέχουν σε λογικές συνθήκες
(συγκρίσεις) και στις πράξεις που ισχύουν στους αντίστοιχους τύπους.

```python
print(days_of_week[1] == "Τρίτη")    # Εμφανίζει: True
print("Μεγάλη " + days_of_week[-3])  # Εμφανίζει: Μεγάλη Παρασκευή
```

## Εγγραφή - Τροποποίηση
Το ίδιο κάνω και για να αλλάξω ένα στοιχείο της λίστας:
```python
days_of_week[5] = "Σάββατο"  # Προσέξτε: Μονό = γιατί κάνω ανάθεση τιμής
print(days_of_week[5])       # Εμφανίζει: Σάββατο
```
