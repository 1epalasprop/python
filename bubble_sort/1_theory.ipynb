# Ταξινόμηση ευθείας ανταλλαγής ή ταξινόμηση φυσαλίδας
Η ταξινόμηση είναι μια λειτουργία που χρησιμοποιείται συχνά στην Πληροφορική.
Κάποιοι γνωστοί αλγόριθμοι ταξινόμησης είναι οι εξής:
- γρήγορη ταξινόμηση
- ταξινόμηση με επιλογή
- ταξινόμηση ευθείας ανταλλαγής ή φυσαλίδα
  
Η **ταξινόμηση φυσαλίδας** είναι ένας αργός αλλά σχετικά εύκολος αλγόριθμος ταξινόμησης.

Ο κώδικας που ταξινομεί σε **αύξουσα σειρά** μία λίστα Α η οποία έχει Ν στοιχεία είναι ο εξής:

```
Ν = len( Α )

   for j in range( N–1 , i , –1 ):
      if Α[ j ] < Α[ j–1 ] :      
         Α[ j ] , Α[ j–1 ] = Α[ j–1 ] , Α[ j ]
```

Έστω η παρακάτω λίστα:
A= [12, 3, -6, 78, -32, 5]

Ο **αλγόριθμος φυσαλίδας** που ταξινομεί τη λίστα αυτή φαίνεται παρακάτω.

````python
A= [12, 3, -6, 78, -32, 5] 
N = len( A )
for i in range( N-1 ):
    for j in range( N-1 , i , -1 ):
        if A[ j ] < A[ j-1 ] :
            A[ j ] , A[ j-1 ] = A[ j-1 ] , A[ j ]
print(A)
````
