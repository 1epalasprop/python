# Εκφώνηση

Ένα ξενοδοχείο ακολουθεί διαφορετική πολιτική χρέωσης των δωματίων ανάλογα με
τον αριθμό των ημερών που θα μείνει κάποιος και ανάλογα την τουριστική περίοδο.
Η χρέωση φαίνεται στον παρακάτω πίνακα:

|       Ημέρες       | Χαμηλή Περίοδος | Υψηλή Περίοδος |
|:------------------:|:---------------:|:--------------:|
|        1 - 3       |       40€       |       70€      |
|        4 - 7       |       35€       |       60€      |
| Περισσότερες από 7 |       30€       |       50€      |

Επίσης, αν το ποσό που θα πληρώσει ο πελάτης ξεπερνά τα 300€, γίνεται *έκπτωση*
10%.

Να γράψετε πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει τον αριθμό ημερών διαμονής.
2. Διαβάζει την τουριστική περίοδο (Χ για χαμηλή και Υ για Υψηλή)
3. Εμφανίζει το ποσό που πρέπει να πληρώσει ο πελάτης.
