# Εκφώνηση

Οι οδηγοί που επιθυμούν να διέρχονται τα διόδια από τις διαδρομές με τα
αυτόματα συστήματα χρησιμοποιούν ειδικές συσκευές (πομποδέκτες / e-pass).
Η χρέωση των διοδίων, ανάλογα με τον τύπο οχήματος γίνεται ως εξής:

| Όχημα                                  | Χρεωση (€) |
|----------------------------------------|:----------:|
| Μοτοποδήλατα (κατηγορία 1)             |    1.25    |
| ΙΧ (κατηγορίες 2, 3 και 4)             |    2.50    |
| Μικρά και μεσαία φορτηγά (κατηγορία 5) |    6.25    |
| Μεγάλα φορτηγά (κατηγορία 6)           |    10.00   |


Το σύστημα διοδίων διαβάζει από τον πομποδέκτη το είδος του οχήματος και το
υπόλοιπο των χρημάτων. Αν τα χρήματα επαρκούν, τότε οι μπάρες ανοίγουν και το
όχημα περνάει διαφορετικά όχι.

Να γράψετε πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει από τον χρήστη την κατηγορία του οχήματος.
2. Διαβάζει το υπόλοιπο των χρημάτων του οχήματος στον πομποδέκτη.
3. Αν το υπόλοιπο είναι μεγαλύτερο από τη χρέωση για την κατηγορία του οχήματος,
   τότε:
   - Εμφανίζει το μήνυμα "Περνάτε"
   - Εμφανίζει το νέο υπόλοιπο (αφού αφαιρέσει τη χρέωση)
   <br>Διαφορετικά:
   - Εμφανίζει το μήνυμα "Το υπόλοιπό σας δεν επαρκεί"

Παράδειγμα εκτέλεσης:
``` 
Δώσε κατηγορία οχήματος: 2
Δώσε το υπόλοιπο των χρημάτων: 8
Περνάτε
Το νέο σας υπόλοιπο είναι 5.50€

Δώσε κατηγορία οχήματος: 6
Δώσε το υπόλοιπο των χρημάτων: 5
Το υπόλοιπό σας δεν επαρκεί
```
