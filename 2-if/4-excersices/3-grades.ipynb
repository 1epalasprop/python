Ο χαρακτηρισμός της φοίτησης των μαθητών γίνεται με βάση τη βαθμολογία τους
σύμφωνα με τον παρακάτω πίνακα:

|    Βαθμός   | Χαρακτηρισμός |
|:-----------:|:-------------:|
|   0 - 9.4   |  ΑΠΟΡΡΙΠΤΕΤΑΙ |
|  9.5 - 15.4 |      ΚΑΛΑ     |
| 15.5 - 17.4 |   ΠΟΛΥ ΚΑΛΑ   |
|  17.5 - 20  |     ΑΡΙΣΤΑ    |

Να γράψετε πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει από τον χρήστη το όνομα ενός μαθητή.
2. Διαβάζει από τον χρήστη το βαθμό του μαθητή.
3. Εμφανίζει τον χαρακτηρισμό της φοίτησης με μήνυμα: "Ο/Η *όνομα_μαθητή* με
   βαθμό *ο_βαθμός_του_μαθητή* έχει χαρακτηρισμό: *ο_χαρακτηρισμός*"
4. Σε περίπτωση που ο βαθμός δεν είναι στην κλίμακα 0 - 20, να εμφανίζεται
   μήνυμα: "Άκυρη βαθμολογία"

Παράδειγμα εκτέλεσης:
```
Δώσε όνομα μαθητή: Σοφία
Δώσε το βαθμό: 18.2
Ο/Η Σοφία με βαθμό 18.2 έχει χαρακτηρισμό: ΑΡΙΣΤΑ

Δώσε όνομα μαθητή: Νίκος
Δώσε το βαθμό: 22
Άκυρη βαθμολογία
```
