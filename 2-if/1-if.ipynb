# Δομή Επιλογής

Η **δομή επιλογής** χρησιμοποιείται όταν θέλουμε να εκτελεστεί μια ακολουθία
εντολών μόνον εφόσον εκπληρώνεται μια συγκεκριμένη συνθήκη. Μας δίνει τη 
δυνατότητα επίλυσης σύνθετων προβλημάτων δημιουργώντας στον αλγόριθμο λογικά
μονοπάτια, εξετάζοντας απλά λογικά ερωτήματα για την εκτέλεση ή όχι μιας ομάδας
εντολών.

# Απλή Δομή Επιλογής

Η **απλή δομή επιλογής** χρησιμοποιείται όταν έχουμε **μία** μόνο επιλογή.

Η σύνταξή της είναι:
```python
if <συνθήκη>:  # η εντολή τελείωνει με άνω-κάτω τελεία
    εντολή_1   # οι εντολές που ακολουθούν "μπαίνουν"
    εντολή_2   # μέσα κατά 4 κενά (ή ένα χαρακτήρα TAB)
    ...        # δημιουργώντας μία "εσοχή".
    εντολή_ν
εντολή_εκτός_if  # Οι εντολές που δεν επηρεάζονται από την if
                 # δεν βρίσκονται σε εσοχή.
```

Η εκτέλεση της εντολής αρχίζει ελέγχοντας τη συνθήκη ελέγχου (`<συνθήκη>`) και
εάν είναι αληθής (True) εκτελείται η ομάδα εντολών που ακολουθεί την `if` και
έχουν εσοχή και στη συνέχεια οι εντολές μετά το τέλος της `if`. Εάν η 
`<συνθήκη>` είναι ψευδής (False), τότε η ροή του προγράμματος θα "προσπεράσει"
τις εντολές που βρίσκονται σε εσοχή και θα συνεχίσει με τις εντολές μετά το 
τέλος της `if`.

# Παραδείγματα

Γράψτε ένα πρόγραμμα το οποίο διαβάζει ένα ακέραιο αριθμό. Αν ο αριθμός είναι
μεγαλύτερος από το 50, τότε αυξήστε τον κατά 10. Εμφανίστε στο τέλος τον αριθμό.

````python
num = int(input("Δώσε έναν αριθμό:"))
if num > 50:
    num = num + 10
print("Το αποτέλεσμα είναι:", num)
````

Γράψτε ένα πρόγραμμα το οποίο διαβάζει έναν ακέραιο αριθμό. Αν ο αριθμός είναι
μικρότερος ή ίσος από 100, μειώστε τον κατά 5. Εμφανίστε στο τέλος τον αριθμό.

````python
arithmos = int(input("Δώσε έναν αριθμό:"))
if arithmos <= 100:
    arithmos = arithmos - 5
print("Το αποτέλεσμα είναι:", arithmos)
````

Γράψτε ένα πρόγραμμα το οποίο θα διαβάζει έναν βαθμό. Στη συνέχεια θα ελέγχει
αν ο βαθμός είναι μεταξύ 0 και 20. Αν είναι, θα εμφανίζει μήνυμα "Έγκυρος
βαθμός".

````python
vathmos = float(input("Δώσε βαθμό:"))
if vathmos >= 0  and vathmos <= 20:
    print("Έγκυρος βαθμός.")
````

Γράψτε ένα πρόγραμμα το οποίο θα διαβάζει ένα βαθμό στην κλίμακα Α-Β-Γ.
Στη συνέχεια θα ελέγχει αν ο βαθμός ισούται με "Α", "Β", ή "Γ", Αν η συνθήκη
είναι Αληθής, τότε θα εμφανίζει μήνυμα "Έγκυρος βαθμός".

````python
vathmos = input("Δώσε βαθμό (Α/Β/Γ):")
if vathmos == "Α" or vathmos == "Β" or vathmos == "Γ":
    print("Έγκυρος βαθμός.")
````
