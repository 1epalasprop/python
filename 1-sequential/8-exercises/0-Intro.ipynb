# Υπολογισμός Μέσου Όρου

## Ορισμός
Ο μέσος όρος είναι μια στατιστική μέτρηση που υπολογίζει την κεντρική τάση ενός
συνόλου αριθμών. Υπολογίζεται προσθέτοντας όλους τους αριθμούς και διαιρώντας
το άθροισμα με τον αριθμό των στοιχείων.

## Οδηγίες
1. Συγκέντρωσε όλα τα αριθμητικά δεδομένα που θέλεις να υπολογίσεις τον μέσο όρο.
1. Πρόσθεσε όλους τους αριθμούς μαζί.
1. Δίαιρεσε το συνολικό άθροισμα με τον αριθμό των δεδομένων.

## Παράδειγμα
Ας υποθέσουμε ότι έχεις τους εξής βαθμούς: 8, 7, 9, 6, 10.
1. Πρόσθεσε τους βαθμούς: 8 + 7 + 9 + 6 + 10 = 40.
2. Δίαιρεσε το άθροισμα με τον αριθμό των βαθμών: 40 ÷ 5 = 8.

**Συμπέρασμα**: Ο μέσος όρος των βαθμών είναι 8.

```python
athroisma = 8 + 7 + 9 + 6 + 10
mesos_oros = athroisma / 5  # 5: το πλήθος των αριθμών στο άθροισμα.
print(mesos_oros)
```

# Υπολογισμός Έκπτωσης

## Ορισμός
Η έκπτωση είναι η μείωση της τιμής ενός προϊόντος ή υπηρεσίας, συνήθως
εκφρασμένη ως ποσοστό. Χρησιμοποιείται συχνά σε πωλήσεις και προσφορές για να
προσελκύσει πελάτες.

## Οδηγίες
1. Καθόρισε την αρχική τιμή του προϊόντος.
2. Υπολόγισε το ποσοστό της έκπτωσης.
3. Πολλαπλασίασε την αρχική τιμή με το ποσοστό της έκπτωσης (σε δεκαδική μορφή).
4. Αφαίρεσε την έκπτωση από την αρχική τιμή για να βρεις την τελική τιμή.

## Παράδειγμα
Ας υποθέσουμε ότι η αρχική τιμή ενός προϊόντος είναι 100 ευρώ και η έκπτωση είναι 20%.

1. Αρχική τιμή: 100 ευρώ.
2. Ποσοστό έκπτωσης: 20% (σε δεκαδική μορφή: 0,20).
3. Υπολόγισε την έκπτωση: 100 × 0,20 = 20 ευρώ.
4. Υπολόγισε την τελική τιμή: 100 - 20 = 80 ευρώ.

**Συμπέρασμα:** Η τελική τιμή του προϊόντος μετά την έκπτωση είναι 80 ευρώ.

Εναλλακτικά, μπορούμε να υπολογίσουμε την έκτπωση χρησιμοποιώντας τον τύπο:
$$ t = a - a \cdot e / 100 $$
όπου $t$ είναι η τελική τιμή, $a$ η αρχική τιμή και $e$ το ποσοστό της έκτπωσης.

Το παραπάνω παράδειγμα γίνεται:
$$ t = 100 - 100 \cdot 20 / 100 = 100 - 2000 / 100 = 100 - 20 = 80 $$

# Υπολογισμός Αύξησης

**Ορισμός:**
Η αύξηση είναι η προσθήκη ενός ποσού σε μια αρχική τιμή. Στην περίπτωση του ΦΠΑ
(Φόρος Προστιθέμενης Αξίας), υπολογίζεται ως ποσοστό της αρχικής τιμής και
προστίθεται στην τιμή αυτή.

## Οδηγίες:
1. Καθόρισε την αρχική τιμή του προϊόντος ή υπηρεσίας.
2. Υπολόγισε το ποσοστό του ΦΠΑ.
3. Πολλαπλασίασε την αρχική τιμή με το ποσοστό του ΦΠΑ (σε δεκαδική μορφή).
4. Πρόσθεσε την αύξηση (ΦΠΑ) στην αρχική τιμή για να βρεις την τελική τιμή.

## Παράδειγμα:
Ας υποθέσουμε ότι η αρχική τιμή ενός προϊόντος είναι 100 ευρώ και ο ΦΠΑ είναι 24%.

1. Αρχική τιμή: 100 ευρώ.
2. Ποσοστό ΦΠΑ: 24% (σε δεκαδική μορφή: 0,24).
3. Υπολόγισε την αύξηση (ΦΠΑ): 100 × 0,24 = 24 ευρώ.
4. Υπολόγισε την τελική τιμή: 100 + 24 = 124 ευρώ.

**Συμπέρασμα:** Η τελική τιμή του προϊόντος μετά την αύξηση του ΦΠΑ είναι 124 ευρώ.

Εναλλακτικά, μπορούμε να υπολογίσουμε την αύξηση χρησιμοποιώντας τον τύπο:
$$ t = a + a \cdot e / 100 $$
όπου $t$ είναι η τελική τιμή, $a$ η αρχική τιμή και $e$ το ποσοστό της αύξησης.

Το παραπάνω παράδειγμα γίνεται:
$$ t = 100 + 100 \cdot 24 / 100 = 100 + 2400 / 100 = 100 + 24 = 124 $$

# πολογισμός Ποσοστού

## Ορισμός
Το ποσοστό είναι μια αναλογία που εκφράζει ένα μέρος σε σχέση με το σύνολο, συνήθως
εκφρασμένο ως ποσοστό (%) του συνόλου.

## Παράδειμα

#### 1. Ξέρουμε το σύνολο και ένα μέρος (π.χ., αριθμός κοριτσιών και θέλουμε το ποσοστό των κοριτσιών).

**Οδηγίες:**
1. Καθόρισε το σύνολο (τον συνολικό αριθμό).
2. Καθόρισε το μέρος (τον αριθμό των κοριτσιών).
3. Χρησιμοποίησε τον τύπο: 
   $$
   \text{Ποσοστό} = \left( \frac{\text{Μέρος}}{\text{Σύνολο}} \right) \times 100
   $$

**Παράδειγμα:**
Ας υποθέσουμε ότι σε μια τάξη υπάρχουν 15 κορίτσια και 25 συνολικά μαθητές.

1. Σύνολο: 25 μαθητές.
2. Μέρος: 15 κορίτσια.
3. Υπολόγισε το ποσοστό: 
   $$
   \text{Ποσοστό} = \left( \frac{15}{25} \right) \times 100 = 60\%
   $$

**Συμπέρασμα:** Το ποσοστό των κοριτσιών είναι 60%.

```python
synolo = 25
koritsia = 15
pososto_koritsia = koritsia / synolo * 100
print("Ποσοστό κοριτσιών:", pososto_koritsia, "%")
```

#### 2. Ξέρουμε το σύνολο και ένα μέρος (π.χ., αριθμός κοριτσιών και θέλουμε το ποσοστό των αγοριών).

**Οδηγίες:**
1. Καθόρισε το σύνολο (τον συνολικό αριθμό).
2. Καθόρισε το μέρος (τον αριθμό των αγοριών).
3. Υπολόγισε τον αριθμό των αγοριών: 
   $$
   \text{Αγόρια} = \text{Σύνολο} - \text{Κορίτσια}
   $$
4. Χρησιμοποίησε τον τύπο: 
   $$
   \text{Ποσοστό} = \left( \frac{\text{Αγόρια}}{\text{Σύνολο}} \right) \times 100
   $$

**Παράδειγμα:**
Ας υποθέσουμε ότι σε μια τάξη υπάρχουν 15 κορίτσια και 25 συνολικά μαθητές.

1. Σύνολο: 25 μαθητές.
2. Μέρος: 15 κορίτσια.
3. Υπολόγισε τον αριθμό των αγοριών: 25 - 15 = 10 αγόρια.
4. Υπολόγισε το ποσοστό: 
   $$
   \text{Ποσοστό} = \left( \frac{10}{25} \right) \times 100 = 40\%
   $$

**Συμπέρασμα:** Το ποσοστό των αγοριών είναι 40%.

```python
synolo = 25
koritsia = 15
agoria = synolo - koritsia
pososto_agoria = agoria / synolo * 100
print("Ποσοστό αγοριών:", pososto_agoria, "%")
```

#### 3. Ξέρουμε τα επιμέρους (π.χ., ξέρουμε το σύνολο αγοριών και κοριτσιών και θέλουμε να υπολογίσουμε το ποσοστό των κοριτσιών).

**Οδηγίες:**
1. Καθόρισε τον αριθμό των κοριτσιών και των αγοριών.
2. Υπολόγισε το σύνολο: 
   $$
   \text{Σύνολο} = \text{Κορίτσια} + \text{Αγόρια}
   $$
3. Χρησιμοποίησε τον τύπο: 
   $$
   \text{Ποσοστό} = \left( \frac{\text{Κορίτσια}}{\text{Σύνολο}} \right) \times 100
   $$

**Παράδειγμα:**
Ας υποθέσουμε ότι υπάρχουν 15 κορίτσια και 10 αγόρια.

1. Κορίτσια: 15.
2. Αγόρια: 10.
3. Υπολόγισε το σύνολο: 15 + 10 = 25.
4. Υπολόγισε το ποσοστό: 
   $$
   \text{Ποσοστό} = \left( \frac{15}{25} \right) \times 100 = 60\%
   $$

**Συμπέρασμα:** Το ποσοστό των κοριτσιών είναι 60%.

```python
koritsia = 15
agoria = 10
synolo = koritsia + agoria
pososto_koritsia = koritsia / synolo * 100
print("Ποσοστό κοριτσιών:", pososto_koritsia, "%")
pososto_agoria = agoria / synolo * 100
print("Ποσοστό αγοριών:", pososto_agoria, "%")
```
