# Εκφώνηση

Το κράτος επιβάλει ένα φόρο προστιθέμενης αξίας (ΦΠΑ, ή VAT στα Αγγλικά) στα
προϊόντα που αγοράζουμε. Έτσι, τα προϊόντα γίνονται ακριβότερα.

Αν ένα προϊόν έχει αρχική τιμή $a$ και το ΦΠΑ είναι 24%, τότε η τελική τιμή του
υπολογίζεται από τον τύπο:
$$ t = a + a \cdot 24 / 100$$

Να γράψετε πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει την αρχική τιμή ενός προϊόντος,
2. Υπολογίζει το ΦΠΑ,
3. Υπολογίζει την τελική τιμή του προϊόντος,
4. Εμφανίζει με κατάλληλα μηνύματα την αρχική τιμή, το ΦΠΑ και την τελική τιμή.

# Επέκταση Εκφώνησης

Το ποσοστό του ΦΠΑ δεν είναι πάντα 24% αλλά αλλάζει για κάποια είδη (π.χ., 
είναι στο 13% για τα τρόφιμα). Αν, λοιπόν, είναι $f$ το ποσοστό του ΦΠΑ,
η τελική τιμή του υπολογίζεται από τον τύπο:
$$ t = a + a \cdot f / 100$$

Να γράψετε πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει την αρχική τιμή ενός προϊόντος,
2. Διαβάζει το ποσοστό του ΦΠΑ,
3. Υπολογίζει το ΦΠΑ,
4. Υπολογίζει την τελική τιμή του προϊόντος,
5. Εμφανίζει με κατάλληλα μηνύματα την αρχική τιμή, το ΦΠΑ και την τελική τιμή.
