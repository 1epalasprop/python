# Εκφώνηση

Το Φαρενάιτ (σύμβολο: °F) είναι μια μονάδα θερμοκρασίας που χρησιμοποιήθηκε ευρέως πριν από τη μετροποίηση. Σήμερα ορίζεται από δύο σταθερά σημεία: τη θερμοκρασία στην οποία παγώνει το νερό, 32°F, και το σημείο βρασμού του νερού, 212°F, και τα δύο σε επίπεδο θάλασσας και τυπική ατμοσφαιρική πίεση. Το διάστημα μεταξύ του σημείου πήξης και του σημείου βρασμού χωρίζεται σε 180 ίσα μέρη.

Η κλίμακα Κελσίου (σύμβολο °C) αντικατέστησε την κλίμακα Φαρενάιτ στις περισσότερες χώρες στα μέσα και στα τέλη του 20ου αιώνα. Σχεδόν όλες οι χώρες σε όλο τον κόσμο χρησιμοποιούν αυτή την κλίμακα, εκτός από εκείνες στις οποίες δεν έχει υιοθετηθεί το μετρικό σύστημα, όπως οι Ηνωμένες Πολιτείες. Ωστόσο, ακόμη και σε χώρες όπως οι Ηνωμένες Πολιτείες, το Κελσίου χρησιμοποιείται ευρέως στην επιστημονική κοινότητα - απλώς δεν χρησιμοποιείται ευρέως στις καθημερινές αναφορές θερμοκρασίας.

Για την μετατροπή από την κλίμακα Κελσίου στη Φαρενάιτ, χρησιμοποιούμε τον τύπο:
$$ F = \frac{9}{5} \cdot C + 32$$
Για την μετατροπή από την κλίμαμα Φαρενάιτ σε Κελσίου, χρησιμοποιούμε τον τύπο:
$$ C = \frac{5}{9} \cdot (F - 32)$$

Να γράψετε πρόγραμμα σε Python το οποίο:
1. Ζητάει από τον χρήστη να εισάγει ένα βαθμό στην κλίμακα **Κελσίου**
2. Μετατρέπει τον βαθμό στην κλίμακα Φαρενάιτ
3. Εμφανίζει το αποτέλεσμα με κατάλληλο μήνυμα στον χρήστη (π.χ. "Οι 5 βαθμοί Κελσίου αντιστοιχούν σε 41 βαθμούς Φαρενάιτ")
