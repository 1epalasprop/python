# Εκφώνηση

Σε ένα κατάστημα κατά την διάρκεια των εκπτώσεων κάνει έκπτωση 15% σε όλα του
τα είδη.

Αν $a$ είναι η αρχική τιμή, η τελική τιμή $t$ υπολογίζεται ως εξής:
$$
t = a - a \cdot 15 / 100
$$

Γράψτε πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει το είδος του προϊόντος που αγοράζει ο πελάτης (π.χ., παντελόνι)
2. Διαβάζει την αρχική τιμή του προϊόντος,
3. Υπολογίζει την τελική τιμή αφαιρώντας την έκπτωση,
4. Εμφανίζει με κατάλληλα μηνύματα την αρχική και την τελική τιμή του προϊόντος,
   π.χ.: "Αρχική τιμή για παντελόνι 100€" "Τελική τιμή 85€"

# Επέκταση Άσκησης

Το κατάστημα κάνει διαφορική έκπτωση την οποία θέλει να καταχωρεί σε κάθε
αγορά. Έτσι, αν $a$ είναι η αρχική τιμή και $e$ το ποσοστό της έκπτωσης, η τελική τιμή $t$
υπολογίζεται ως εξής:
$$
t = a - a \cdot e / 100
$$

Τροποποιήστε κατάλληλα το παραπάνω πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει το είδος του προϊόντος που αγοράζει ο πελάτης (π.χ., παντελόνι)
2. Διαβάζει την αρχική τιμή του προϊόντος,
3. Διαβάζει το ποσοστό της έκπτωσης,
4. Υπολογίζει την τελική τιμή αφαιρώντας την έκπτωση,
5. Εμφανίζει με κατάλληλα μηνύματα την αρχική και την τελική τιμή του προϊόντος,
   π.χ.: "Αρχική τιμή για παντελόνι 100€" "Τελική τιμή 85€"
