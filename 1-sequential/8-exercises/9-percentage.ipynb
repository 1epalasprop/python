# Εκφώνηση #1

Σε μια έρευνα για τις προτιμήσεις των καταναλωτών, ένα σύνολο ατόμων ρωτήθηκαν σχετικά με την αγαπημένη τους γεύση παγωτού. Από αυτά, κάποια άτομα δήλωσαν ότι προτιμούν τη γεύση σοκολάτα.

Γράψε ένα πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει το σύνολο των ατόμων που ρωτήθηκαν.
2. Διαβάζει για πόσους από αυτούς η αγαπημένη τους γεύση είναι σοκολάτα.
3. Υπολογίζει το ποσοστό των ατόμων που προτιμούν τη γεύση σοκολάτα σε σχέση με το σύνολο των συμμετεχόντων στην έρευνα.
4. Εμφανίζει το αποτέλεσμα με κατάλληλο μήνυμα:
   Παράδειγμα: *"Από τα 150 άτομα που ρωτήθηκαν τα 90 δήλωσαν ότι προτιμούν σοκολάτα. Ποσοστό 60.0%"*

**Σημείωση:** Χρησιμοποίησε τον τύπο για τον υπολογισμό του ποσοστού: 
$$
\text{Ποσοστό} = \left( \frac{\text{Μέρος}}{\text{Σύνολο}} \right) \times 100
$$

Υπολογίστε το ποσοστό των ατόμων που **δεν** προτιμούν τη γεύση σοκολάτα.

# Εκφώνηση #2

Σε μια έρευνα για τις προτιμήσεις των καταναλωτών, ένα σύνολο ατόμων ρωτήθηκαν σχετικά με την αγαπημένη τους γεύση παγωτού ανάμεσα σε βανίλια και σοκολάτα. Από αυτά, κάποια άτομα δήλωσαν ότι προτιμούν τη γεύση σοκολάτα.

Γράψε ένα πρόγραμμα σε γλώσσα Python το οποίο:
1. Διαβάζει το σύνολο των ατόμων η αγαπημένη τους γεύση είναι σοκολάτα.
2. Διαβάζει το σύνολο των ατόμων η αγαπημένη τους γεύση είναι βανίλια.
3. Υπολογίζει το ποσοστό των ατόμων που προτιμούν τη γεύση σοκολάτα σε σχέση με το σύνολο των συμμετεχόντων στην έρευνα.
4. Υπολογίζει το ποσοστό των ατόμων που προτιμούν τη γεύση βανίλια σε σχέση με το σύνολο των συμμετεχόντων στην έρευνα.
4. Εμφανίζει το αποτέλεσμα με κατάλληλο μήνυμα:
   Παράδειγμα: *"Από τα 150 άτομα που ρωτήθηκαν τα 90 δήλωσαν ότι προτιμούν σοκολάτα. Ποσοστό 60.0%"*
   Παράδειγμα: *"Από τα 150 άτομα που ρωτήθηκαν τα 60 δήλωσαν ότι προτιμούν βανίλια. Ποσοστό 40.0%"*
