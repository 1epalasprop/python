# Μεταβλητές

Μια μεταβλητή στην Python είναι σαν ένα κουτί που μπορεί να αποθηκεύσει δεδομένα.
Σκεφτείτε τη μεταβλητή ως ένα όνομα που δίνετε σε μια συγκεκριμένη τιμή, όπως
ένας αριθμός ή ένα κείμενο. Για παράδειγμα, αν θέλετε να αποθηκεύσετε την ηλικία
σας, μπορείτε να δημιουργήσετε μια μεταβλητή με το όνομα `age` και να της
δώσετε την τιμή σας, π.χ. `age = 16`. Αυτό σας επιτρέπει να χρησιμοποιείτε αυτή
τη μεταβλητή σε όλο το πρόγραμμά σας, χωρίς να χρειάζεται να θυμάστε την
πραγματική τιμή κάθε φορά. Οι μεταβλητές κάνουν τον κώδικα πιο ευανάγνωστο και
ευκολότερο στη διαχείριση.

Στις μεταβλητές αποθηκεύουμε οποιονδήποτε από τους τύπους δεδομένων που μάθαμε.
Χρησιμοποιούμε τον όρο *"περιεχόμενο"* για να αναφερθούμε στην τιμή που είναι
αποθηκευμένη στη μεταβλητή.

## Κανόνες Ονομάτων Μεταβλητών

Οι κανόνες για τα ονόματα των μεταβλητών στην Python είναι οι εξής:

1. **Έναρξη με γράμμα ή κάτω παύλα:** Τα ονόματα των μεταβλητών πρέπει να
   ξεκινούν με ένα γράμμα (a-z, A-Z) ή με κάτω παύλα (\_).
   Δεν μπορούν να ξεκινούν με αριθμό.
2. **Χρήση γραμμάτων, αριθμών και κάτω παύλας:** Μετά το πρώτο χαρακτήρα, τα
   ονόματα των μεταβλητών μπορούν να περιέχουν γράμματα, αριθμούς (0-9) και
   κάτω παύλα (_).
3. **Διάκριση πεζών-κεφαλαίων:** Η Python διακρίνει πεζά και κεφαλαία γράμματα,
   δηλαδή οι μεταβλητές `age`, `Age` και `AGE` είναι διαφορετικές.
4. **Αποφυγή δεσμευμένων λέξεων:** Δεν μπορούμε να χρησιμοποιούμε δεσμευμένες
   λέξεις της Python (όπως `if`, `else`, `for`, `while`, κ.λπ.) ως ονόματα
   μεταβλητών.
   
Ακολουθώντας αυτούς τους κανόνες, μπορούμε να δημιουργούμε ονόματα μεταβλητών
που είναι κατανοητά και εύκολα στη χρήση.

### Αποδεκτά Ονόματα Μεταβλητών

* `age`
* `student_name`
* `mesos_oros`
* `x`
* `data2024`

### Μη Αποδεκτά Ονόματα Μεταβλητών

* `2age` (ξεκινάει με αριθμό)
* `student-name` (περιέχει παύλα)
* `mesos oros` (περιέχει κενό)
* `x$` (ο χαρακτήρας $ δεν είναι ούτε γράμμα, ούτε αριθμός, ούτε κάτω παύλα)
* `if` (δεσμευμένη λέξη)

## Εκχώρηση Τιμής σε Μεταβλητή

Η εκχώρηση τιμής σε μεταβλητή, γίνεται ως εξής:

    όνομα_μεταβλητής = τιμή
    
Η εκχώρηση τιμής σε μια μεταβλητή στην Python γίνεται με το σύμβολο =. Αυτό το σύμβολο δεν σημαίνει ισότητα, όπως στα μαθηματικά, αλλά εκχώρηση. Δηλαδή, όταν γράφουμε x = 5, αυτό σημαίνει ότι δίνουμε στη μεταβλητή x την τιμή 5. Η μεταβλητή x τώρα “κρατάει” την τιμή 5 και μπορούμε να τη χρησιμοποιήσουμε αργότερα στο πρόγραμμά μας. Είναι σημαντικό να θυμόμαστε ότι το = στην Python δεν συγκρίνει δύο τιμές, αλλά αποθηκεύει την τιμή που βρίσκεται στα δεξιά του στη μεταβλητή που βρίσκεται στα αριστερά του.

### Ταυτόχρονη Εκχώρηση Τιμών σε Πολλές Μεταβλητές

Μπορούμε να συνδυάσουμε εκχωρήσεις τιμών σε μεταβλητές σε μία γραμμή.
Για παράδειγμα αν έχουμε:

    όνομα_μεταβλητής1 = τιμή1
    όνομα_μεταβλητής2 = τιμή2

μπορούμε να γράψουμε:

    όνομα_μεταβλητής1, όνομα_μεταβλητής2 = τιμή1, τιμή2

Δηλαδή, γράφουμε όσα ονόματα μεταβλητών χρειάζεται πριν το σύμβολο της
εκχώρησης (`=`) και μετά τις τιμές που θα πάρουν.

Προσέχουμε τα εξής:
* Τα ονόματα των μεταβλητών *και* οι τιμές χωρίζονται με **κόμμα**
* Όσα ονόματα μεταβλητών έχουμε πριν το `=`, τόσες τιμές πρέπει να έχουμε και
  μετά.

Αυτός ο τρόπος εκχώρησης έχει νόημα όταν οι μεταβλητές έχουν κάποια σχέση.
Παραδείγματα:

* Συντεταγμένες ενός σημείου στο επίπεδο (x, y):

    x, y = 4, -1

* Συντεταγμένες ενός σημείου στο χώρο (x, y, z):

    x, y, z = 5, 2, -3

* Όνομα, επίθετο και ηλικία ενός ατόμου:

    first_name, last_name, age = "John", "Doe", 20

### Αλυσιδωτή Εκχώρηση Τιμής

Αν θέουμε πολλές *διαφορετικές* μεταβλητές να πάρουν **την ίδια** τιμή, τότε
αντί να γράψουμε:

    x = 0
    y = 0
    z = 0

μπορούμε να γράψουμε:

    x = y = z = 0

## Εμφάνιση Τιμής Μεταβλητής

Για να εμφανίσουμε το περιεχόμενο μιας μεταβλητής στην Python, χρησιμοποιούμε
τη συνάρτηση **`print()`**. Αυτή η συνάρτηση παίρνει ως είσοδο τη μεταβλητή
που θέλουμε να εμφανίσουμε και εκτυπώνει την τιμή της στην οθόνη. Για
παράδειγμα, αν έχουμε τη μεταβλητή `x` με τιμή `5`, μπορούμε να γράψουμε
`print(x)` και η τιμή `5` θα εμφανιστεί στην οθόνη.

Η συνάρτηση `print()` είναι πολύ χρήσιμη για τον έλεγχο και την παρακολούθηση
των τιμών των μεταβλητών κατά την εκτέλεση του προγράμματος.

```python
x = 5
print(x)
```

Στην `print()` μπορούμε να συνδυάσουμε μήνυμα σε συμβολοσειρά και όνομα
μεταβλητής χρησιμοποιώντας το κόμμα ως εξής:

    print("συμβολοσειρά", όνομα_μεταβλητής)

```python
x = 10
print("Η τιμή του x είναι:", x)
print("x =", x)
```

```python
# Αριθμητική τιμή
age = 25
print("Η ηλικία είναι:", age)
```

```python
# Λογική τιμή
has_passed = False
print("Έχει περάσει:", has_passed)
```

```python
# Συμβολοσειρά
name = "John"
print("Το όνομα είναι:", name)
```

```python
# Συντεταγμένες
x, y = 5, -1
print("Οι συντεταγμένες είναι:", x, y)
```

```python
# Όνομα, επώνυμο και ηλικά ατόμου
first_name, last_name, age = "John", "Doe", 20
print("Στοιχεία ατόμου", first_name, last_name, age)
```

```python
# Χρώματα συμβόλων τράπουλας
spades = clubs = "Μαύρο"
hearts = diamonds = "Κόκκινο"
print("Το χρώμα για τα 'μπαστούνια' είναι", spades)
print("Το χρώμα για τα 'σπαθιά' είναι", clubs)
print("Το χρώμα για τα 'καρό' είναι", diamonds)
print("Το χρώμα για τις 'κούπες' είναι", hearts)
```
