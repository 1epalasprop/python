# Αριθμητικοί Τελεστές

Ο όρος "τελεστής" είναι συνώνυμο με τον όρο "σύμβολο". Δηλαδή, οι
_Αριθμητικοί Τελεστές_ είναι τα _σύμβολα_ που χρησιμοποιούμε για να κάνουμε
_πράξεις_ μεταξύ αριθμών.

Οι πράξεις είναι οι γνωστές από τα μαθηματικά με τη διαφορά ότι στην Python
χρησιμοποιούμε διαφορετικά σύμβολα για κάποιες.
Επίσης, σε κάθε έκφραση στην οποία υπάρχουν αριθμητικοί τελεστές, ακολουθείτε
η ιεραρχία των πράξεων που φαίνεται στον παρακάτω πίνακα.

| Ιεραρχία | Πράξη                       | Τελεστής |
|:--------:|-----------------------------|:--------:|
|     1    | Υψωση σε δύναμη             |    **    |
|     2    | Πολλαπλασιασμός             |     *    |
|     2    | Διαίρεση                    |     /    |
|     2    | Υπόλοιπο ακέραιας διαιρεσης |     %    |
|     3    | Πρόσθεση                    |     +    |
|     3    | Αφαίρεση                    |     -    |

Όταν θέλουμε να αλλάξουμε την ιεραρχία των πράξεων, χρησιμοποιούμε παρενθέσεις.

# Πράξεις με Μεταβλητές

Όταν έχουμε μεταβλητές στις οποίες έχουμε εκχωρήσει αριθμητικούς τύπους, 
μπορούμε να τις χρησιμοποιήσουμε σαν αριθμούς. Δηλαδή, μπορούμε να κάνουμε
πράξεις με αυτές.

Για παράδειγμα, για να αυξήσω την τιμή που έχει μια μεταβλητή, θα πρέπει να
προσθέσω σ' αυτή  ένα αριθμό. Η εκχώρηση γίνεται ως εξής:

```python
counter = 1
counter = counter + 1
```

Στην 1η γραμμή δημιουργούμε μία μεταβλητή με όνομα `counter` και της εκχωρούμε
το `1`. Επειδή είναι η πρώτη φορά που εμφανίζεται αυτή η μεταβλητή και παίρνει
τιμή, λέμε ότι έχει «αρχική τιμή» `1`.

Στη 2η γραμμή, η Python υπολογίζει πρώτα το δεξί μέλος της εκχώρησης (`counter + 1`)
και μετά το δίνει στη μεταβλητή που βρίσκεται στο αριστερό μέλος της εκχώρησης.
Έτσι, αφού ο `counter` έχει την τιμή `1`, αυτό που γίνεται είναι:

```python
counter = 1 + 1 
counter = 2
```

Δηλαδή, ο `counter` έχει πλέον την τιμή `2`.

Αντίστοιχα λειτουργούν όλοι οι αριθμητικοί τελεστές.

Γνωρίζοντας την πλευρά $a$ του τετραγώνου, μπορούμε να υπολογίσουμε τα εξής μεγέθη:
* Περίμετρος: $P = 4 \cdot a$
* Εμβαδόν: $E = a^2$
* Ακτίνα εγγεγραμμένου κύκλου: $r = \frac{a}{2}$

Ορίστε μια μεταβλητή με όνομα `a` και αρχική τιμή `5`. Στη συνέχεια υπολογίστε
τα παραπάνω μεγέθη και εμφανίστε τα.

````python
a = 5
p = 4 * a
e = a ** 2
r = a / 2
print("Περίμετρος:", p)
print("Εμβαδόν:", e)
print("Ακτίνα εγγεγραμμένου κύκλου:", r)
````
