# Εκφώνηση

Σε ένα αγώνα ρίψης ακοντίου, διεξάγεται ο προκριματικός γύρος με τη συμμετοχή
8 αθλητών. Στην τελική φάση προκρίνονται όσοι αθλητές επιτύχουν επίδοση άνω των
80 μέτρων.

Να γράψετε πρόγραμμα ώστε:
1. Να διαβάζει για κάθε έναν αθλητή την καλύτερη επίδοσή του.
2. Αν η επίδοση είναι άνω των 80 μέτρων να τον μετράει σε αυτούς που προκρίνονται.
3. Να εμφανίζει πόσοι αθλητές προκρίθηκαν.
4. Να υπολογίζει και να εμφανίζει το ποσοστό των αθλητών που προκρίθηκαν.

# Λύση

## Ζητούμενο 1

Θα πρέπει να διαβάσουμε διαδοχικά την επίδοση 8 αθλητών. Γνωρίζουμε το πλήθος
τους, άρα θα χρησιμοποιήσουμε τη `for`.

```python
for i in range(8):
    epidosi = float(input("Επίδοση αθλητή: "))
```

# Ζητούμενο 2

Θα χρειαστούμε μία μεταβλητή την οποία θα αυξάνουμε όταν ο αθλητής περάσει το
όριο. Η αύξηση αυτή θα γίνεται υπό συνθήκη, δηλαδή, *αν η επίδοση είναι 
μεγαλύτερη των 80 μέτρων, τότε μέτρα τον αθλητή*.

```python
prokrisi = 0  # Ξεκινώντας, κανείς δεν έχει προκριθεί.
for ...:
    epidosi = ...
    if epidosi > 80:
        prokrisi = prokrisi + 1
```

# Ζητούμενο 3

Για να εμφανίσουμε το πλήθος των αθλητών που προκρίθηκαν, θα πρέπει να το
κάνουμε μετά τη for έτσι ώστε να έχει ολοκληρωθεί η καταμέτρηση.

```python
prokrisi = 0  # Ξεκινώντας, κανείς δεν έχει προκριθεί.
for ...:
    epidosi = ...
    if ...:
        ...

print("Προκρίθηκαν", prokrisi, "αθλητές.")
```

# Ζητούμενο 4

Ο υπολογισμός του ποσοστού γίνεται με την πράξη:
```
υποσύνολο / σύνολο * 100
```

Για εμάς το υποσύνολο είναι οι αθλητές που προκρίθηκαν και το σύνολο 8.
```python

for ...:
    epidosi = ...

        ...

```
