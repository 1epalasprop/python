# Εκφώνηση

Η ομάδα «αετόπουλα» των προσκόπων του Δήμου Ασπροπύργου αποφάσισε να διεξάγει έρανο για φιλανθρωπικούς σκοπούς. Οι πρόσκοποι πρέπει να πουλήσουν κουτιά με μπισκότα και ο Δήμος ως κίνητρο συμμετοχής παρέχει τα έξοδα διαμονής σε γνωστή κατασκήνωση της Χαλκιδικής για τον πρόσκοπο που θα πουλήσει τα περισσότερα κουτιά. Γράψτε πρόγραμμα το οποίο:

1. Διαβάζει το πλήθος των προσκόπων που συμμετέχουν στον έρανο. Ο αλγόριθμος πρέπει να επιβεβαιώνει ότι το πλήθος είναι θετικός αριθμός, διαφορετικά ζητά εκ νέου από το χρήστη να δώσει το πλήθος των συμμετεχόντων μέχρι να δώσει έγκυρο αριθμό.
2. Για κάθε έναν πρόσκοπο διαβάζει το επώνυμό του και τα κουτιά που πούλησε.
3. Εμφανίζει το επώνυμο του προσκόπου που πούλησε τα περισσότερα κουτιά.
4. Εμφανίζει το συνολικό ποσό που συγκεντρώθηκε, δεδομένου ότι κάθε κουτί από μπισκότα πωλείται προς 5€.
