# Εύρεση του Μεγαλύτερου/Μικρότερου (max/min) Αριθμού

Για να βρούμε τον μεγαλύτερο αριθμό, σε μια ακολουθία *άγνωστων* αριθμών,
ακολουθούμε την εξής διαδικασία *όταν ο αριθμός επαναλλήψεων είναι γνωστός**:

1. Διαβάζουμε τον 1ο αριθμό και τον αποθηκεύουμε στη μεταβλητή που θα έχει στο τέλος τη μεγαλύτερη τιμή
2. Χρησιμοποιούμε τη δομή επανάληψης for
3. Στη συνάρτηση range βάζουμε τους κατάλληλος αριθμούς ώστε να φτιάξουμε τον κατάλληλο αριθμό επαναλήψεων. **Προσοχή**: Θα πρέπει να βάλουμε μία επανάληψη λιγότερη γιατί έχουμε ήδη διαβάσει έναν αριθμό στο βήμα 1
4. Σε κάθε επανάληψη της for διαβάζουμε έναν αριθμό
5. συγκρίνουμε αυτό τον αριθμό αν είναι μεγαλύτερος από τον μέγιστο που έχουμε
6. Αν ναι, θέτουμε την τιμή αυτή για μέγιστο, διαφορετικά συνεχίζουμε στην επόμενη επανάληψη
7. Στο τέλος της διαδικασίας, η μεταβλητή από το βήμα 1 θα έχει τον μεγαλύτερο αριθμό από όσους διαβάσαμε.

Για να βρούμε τον μικρότερο αριθμό, αλλάζει μόνο ο τελεστής της σύγκρισης στο βήμα 5.

## Παράδειγμα

Διαβάστε 5 αριθμούς από τον χρήστη και βρείτε το μεγαλύτερο αριθμό.
Ο χρήστης δίνει σ' αυτό το παράδειγμα, την ακολουθία `7 3 12 0 9`.

```python
maxi = int(input("Δώσε έναν αριθμό: "))  # Πάντα πριν τη for
for i in range(4):  # Μία επανάληψη λιγότερη γιατί έχουμε διαβάσει έναν αριθμό
    n = int(input("Δώσε έναν αριθμό: "))
    if n > maxi:  # Αν ο αριθμός n είναι μεγαλύτερος από τον maxi
        maxi = n  # Κάνε τον maxi ίσο με τον n
print("Ο μεγαλύτερος αριθμός είναι:", maxi)
```
