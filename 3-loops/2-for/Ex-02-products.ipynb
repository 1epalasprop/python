# Γινόμενο Ακολουθίας Αριθμών

Για να υπολογίσουμε το γινόμενο μίας *γνωστής* ακολουθίας αριθμών, ακολουθούμε
την παρακάτω διαδικασία:

1. Δίνουμε σε μια μεταβλητή (έστω, product, ginomeno, ...) την αρχική τιμή 1
2. Χρησιμοποιούμε τη δομή επανάληψης for
3. Στη συνάρτηση range βάζουμε τους κατάλληλους αριθμούς για να φτιάξουμε την ακολουθία
4. Σε κάθε επανάληψη της for πολλαπλασιάζουμε τη μεταβλητή της με τη product, ginomeno, ή όπως την ονομάσατε
5. Στο τέλος της επανάληψης η μεταβλητή του βήματος 1 θα έχει την τελική τιμή.

## Παράδειγμα

Γράψτε ένα πρόγραμμα με τη χρήση της δομής επανάληψης **for** που θα υπολογίζει
και θα εμφανίζει το γινόμενο όλων των αριθμών από το 1 έως το 10. Δηλαδή, το γινόμενο:

`1 * 2 * 3 * ... * 10`

Το αποτέλεσμα που πρέπει να πάρετε από το πρόγραμμα είναι: **3628800**

````python
product = 1
for i in range(1, 11):
    product = product * i
print(product)
````
