# Εύρεση του Μεγαλύτερου/Μικρότερου (max/min) Αριθμού

Για να βρούμε τον μεγαλύτερο αριθμό, σε μια ακολουθία *άγνωστων* αριθμών,
ακολουθούμε την εξής διαδικασία:

1. Δίνουμε στη μεταβλητή *`maxi`* (που στο τέλος θα έχει τον μεγαλύτερο αριθμό)
   μια πολύ μικρή τιμή, π.χ., το `-9e10`
1. Διαβάζουμε τον 1ο αριθμό σε μία άλλη μεταβλητή (έστω `num`)
1. Στη συνθήκη της `while` ελέγχουμε αν η μεταβλητή `num` με τον φρουρό
1. Μέσα στη `while` συγκρίνουμε αυτό τον αριθμό αν είναι μεγαλύτερος από τον
   μέγιστο που έχουμε
1. Αν ναι, θέτουμε την τιμή αυτή για μέγιστο, διαφορετικά συνεχίζουμε στην
   επόμενη επανάληψη
1. Στο τέλος της `while` διαβάζουμε ξανά έναν αριθμό
1. Στο τέλος της διαδικασίας, η μεταβλητή από το βήμα 1 θα έχει τον μεγαλύτερο
   αριθμό από όσους διαβάσαμε.

Για να βρούμε τον μικρότερο αριθμό

1. Στο βήμα 1 δίνουμε μια πολύ μεγάλη τιμή, π.χ., το `9e10`
2. Αλλάζει ο τελεστής της σύγκρισης στο βήμα 5.

## Παράδειγμα

Διαβάστε αριθμούς από τον χρήστη *μέχρι να δώσει το -99999 (φρουρός)* και 
βρείτε το μεγαλύτερο αριθμό από όσους έδωσε.

Ο χρήστης δίνει σ' αυτό το παράδειγμα, την ακολουθία `7 3 12 0 9 -99999`.

```python
maxi = -9e10  # Βήμα 1
num = int(input("Δώσε έναν αριθμό: "))  # Βήμα 2
while num != -99999:  # Βήμα 3
    if num > maxi:  # Βήμα 4. Αν ο αριθμός n είναι μεγαλύτερος από τον maxi
        maxi = num  # Βήμα 5. Κάνε τον maxi ίσο με τον num
    num = int(input("Δώσε έναν αριθμό: "))  # Βήμα 6
print("Ο μεγαλύτερος αριθμός είναι:", maxi)  # Βήμα 7
```
