# Άθροισμα Ακολουθίας Αριθμών

Για να υπολογίσουμε το άθροισμα μίας *άγνωστης* ακολουθίας αριθμών, ακολουθούμε
την παρακάτω διαδικασία:

1. Δίνουμε μία μεταβλητή (έστω, suma, athroisma, ...) αρχική τιμή 0
2. Διαβάζουμε την 1η τιμή και τη δίνουμε σε μία μεταβλητή (έστω, timi, num, ...)
3. Χρησιμοποιούμε τη δομή επανάληψης **while**
4. Στη while προσέχουμε να βάλουμε σωστή τη συνθήκη ώστε να τερματίσει κάποια στιγμή
5. Σε κάθε επανάληψη της while προσθέτουμε την τιμη που διαβάσαμε στη suma
6. Μέσα στο βρόχο της επανάληψης θα πρέπει να διαβάσουμε πάλι μία τιμή
7. Στο τέλος της επανάληψης η μεταβλητή του βήματος 1 θα έχει την τελική τιμή.

Τα βήματα 5 & 6 μπορεί να αλλάξουν σειρά ανάλογα με το πρόβλημα που λύνουμε.

## Παράδειγμα 1

Διαβάστε από το πληκτρολόγιο και αθροίστε τιμές. Η διαδικασία σταματάει όταν δωθεί η τιμή 0.

* Δε γνωρίζουμε πόσες τιμές θα διαβάσουμε μέχρι να δωθεί το 0, γι' αυτό θα χρησιμοποιήσουμε τη while
* Η συνθήκη τερματισμού είναι η τιμή που διαβάζουμε, άρα αυτή θα πρέπει να είναι στη συνθήκη του while

### Εξάσκηση

* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε τις τιμές `12 6.5 3.5 0`
* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε τις τιμές `-10 0 10`
* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε την τιμή `0`


```python
suma = 0  # Βήμα 1
timi = float(input("Δώσε την τιμή: "))  # Βήμα 2
while timi != 0:  # Βήμα 3 & 4
    suma = suma + timi  # Βήμα 5
    timi = float(input("Δώσε την τιμή: "))  # Βήμα 6
print(suma)  # Βήμα 7
```

## Παράδειγμα 2

Διαβάστε από το πληκτρολόγιο και αθροίστε τιμές μέχρι το άθροισμα να ξεπεράσει το 1000.

* Δε γνωρίζουμε πόσες τιμές θα διαβάσουμε μέχρι να ξεπεράσουμε το όριο (1000) γι' αυτό θα χρησιμοποιήσουμε τη δομή while. 
* Η συνθήκη τερματισμού μας είναι το άθροισμα, άρα αυτό θα πρέπει να μπει στη συνθήκη του while.

### Εξάσκηση

* Τρέξε των κώδικα στο παρακάτω κελι και δώσε τις τιμές `500 250 300`
* Τρέξε των κώδικα στο παρακάτω κελι και δώσε τις τιμές `-500 500 300 800`
* Τρέξε των κώδικα στο παρακάτω κελι και δώσε τις τιμές `1100`

```python
suma = 0  # Βήμα 1
timi = float(input("Δώσε την τιμή: "))  # Βήμα 2
suma = suma + timi  # Αφού έχουμε διαβάσει μία τιμή θα πρέπει να τη βάλουμε στο άθροισμα
# Αυτό το κάνουμε επειδή το άθροισμα θα μπει στη συνθήκη τερματισμού.
while suma <= 1000:  # Βήμα 3 & 4. Συνθήκη τερματισμού.
    timi = float(input("Δώσε την τιμή: "))  # Βήμα 6. Δε ξεχνάμε να διαβάσουμε μία νέα τιμή.
    suma = suma + timi  # Βήμα 5. Δε ξεχνάμε να προσθέσουμε τη νέα τιμή.
print(suma)  # Βήμα 7
```
