# Ο Βρόχος `while`

Στη γενική μορφή του ένας βρόχος `while` είναι ως εξής:
```
1. x = input()
2. while η_συνθήκη_που_περιλαμβάνει_το_x_είναι_αληθής:
3.    εντολή_1
4.    εντολή_2
      ...
5.    x = input()
```

H `η_συνθήκη_που_περιλαμβάνει_το_x_είναι_αληθής` ελέγχει το `x` με κάποια τιμή
που, συνήθως, την ονομάζουμε *φρουρό*.

Θέλουμε, λοιπόν:
* Η μεταβλητή που συγκρίνουμε με τον φρουρό να έχει πάρει τιμή *πριν* από τη
  while (γραμμή 1)
* Η μεταβλητή που συγκρίνουμε με τον φρουρό να έχει πάρει τιμή *μέσα* στη while
  (συνήθως στο τέλος, γραμμή 5)
  
# Συνθήκες Τερματισμού (Φρουροί)
Με άλλα λόγια, θέλουμε να γράψουμε ένα βρόχο ο οποίος εκτελείτε *όσο η 
μεταβλητή της συνθήκη τερματισμού είναι **διαφορετική** από τον φρουρό*. Αυτό
σημαίνει ότι όταν η μεταβλητή γίνει ίση με τον φρουρό, ο βρόχος θα τερματίσει.

## Παράδειγμα 1

Διαβάζετε αριθμούς μέχρι να δωθεί η τιμή 0. Στο τέλος εμφανίστε το πλήθος
των αριθμών που διαβάσατε.

Σ' αυτό το παράδειγμα, ο αριθμός φρουρός είναι το 0. Δίνουμε τους αριθμούς `9 4 10 8 0`

````python
count = 0  # Μετρητής αριθμών
num = float(input("Δώσε έναν αριθμό: "))
while num != 0:  # Όσο η μεταβλητή num είναι διαφορετική από τον φρουρό
    count = count + 1  # Ο αριθμός δεν ήταν ο φρουρός. Μέτρα τον!
    num = float(input("Δώσε έναν αριθμό: "))  # Πρέπει να διαβάσουμε έναν νέο αριθμό.

print("Πλήθος αριθμών:", count)
````

## Παράδειγμα 2

Διαβάστε ονόματα μαθητών μέχρι να δωθεί ως όνομα η λέξη "Τέλος". Στο τέλος
εμφανίστε το πλήθος των ονομάτων που διαβάσατε.

Σ' αυτό το παράδειγμα η λέξη φρουρός είναι "Τέλος". Δίνουμε τα ονόματα:
`Δίας Ηρα Αθηνά Ερμής Τέλος`

````python
count = 0 # Μετρητής ονομάτων
name = input("Δώσε το όνομα: ")
while name != "Τέλος":  # Όσο η μεταβλητή name είναι διαφορετική από τον φρουρό
    count = count + 1  # Το όνομα δεν ήταν ο φρουρός. Μέτρα το!
    name = input("Δώσε το όνομα: ")  # Πρέπει να διαβάσουμε ένα νέο όνομα

print("Πλήθος ονομάτων: ", count)
````
