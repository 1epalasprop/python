# Γινόμενο Ακολουθίας Αριθμών

Για να υπολογίσουμε το γινόμενο μίας *άγνωστης* ακολουθίας αριθμών, ακολουθούμε
την παρακάτω διαδικασία:

1. Δίνουμε σε μια μεταβλητή (έστω, product, ginomeno, ...) την αρχική τιμή 1
1. Διαβάζουμε την 1η τιμή και τη δίνουμε σε μια μεταβλητή (έστω timi, num, ...)
1. Χρησιμοποιούμε τη δομή επανάληψης `while`
1. Στη `while` προσέχουμε να βάλουμε σωστή τη συνθήκη ώστε να τερματίσει κάποια στιγμή
1. Σε κάθε επανάληψη της `while` πολλαπλασιάζουμε τη τιμή που διαβάσαμε με τη `product`
1. Διαβάζουμε μία νέα τιμή στη μεταβλητή `timi`
1. Στο τέλος της επανάληψης η μεταβλητή του βήματος 1 θα έχει την τελική τιμή.

## Παράδειγμα 1

Διαβάστε από το πληκτρολόγιο και υπολογίστε το γινόμενό τους. Η διαδικασία σταματάει όταν δωθεί η τιμή 0.

* Δε γνωρίζουμε πόσες τιμές θα διαβάσουμε μέχρι να δωθεί το 0, γι' αυτό θα χρησιμοποιήσουμε τη while
* Η συνθήκη τερματισμού είναι η τιμή που διαβάζουμε, άρα αυτή θα πρέπει να είναι στη συνθήκη του while

### Εξάσκηση

* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε τις τιμές `12 6.5 3.5 0`
* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε τις τιμές `-10 0 10`
* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε την τιμή `0`


```python
product = 1  # Βήμα 1
timi = float(input("Δώσε μία τιμή: "))  # Βήμα 2
while timi != 0:  # Βήμα 3 & 4
    product = product * timi  # Βήμα 5
    timi = float(input("Δώσε μία τιμή: "))  # Βήμα 6
print(product)  # Βήμα 7
```

## Παράδειγμα 2

Διαβάστε από το πληκτρολόγιο και αθροίστε τιμές μέχρι το άθροισμα να ξεπεράσει το 1000.

* Δε γνωρίζουμε πόσες τιμές θα διαβάσουμε μέχρι να ξεπεράσουμε το όριο (1000) γι' αυτό θα χρησιμοποιήσουμε τη δομή while. 
* Η συνθήκη τερματισμού μας είναι το γινόμενο, άρα αυτό θα πρέπει να μπει στη συνθήκη του while.

### Εξάσκηση

* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε τις τιμές `12 12 10`
* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε τις τιμές `10 101`
* Τρέξε τον κώδικα στο παρακάτω κελί και δώσε την τιμή `-10 -101`


```python
product = 1
timi = float(input("Δώσε μία τιμή: "))
product = product * timi
while product <= 1000:
    timi = float(input("Δώσε μία τιμή: "))
    product = product * timi
print(product) 
```
